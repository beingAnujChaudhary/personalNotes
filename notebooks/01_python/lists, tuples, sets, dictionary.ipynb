#  Lists (The Ultimate Container)

## 1. What are Lists?

A **List** is a built-in Python data structure used to store ordered, mutable collections of data. They are defined by square brackets `[]`.

* **Python List:** An array of **pointers** pointing to scattered objects. Highly flexible, but slow and memory-heavy.
* **NumPy Array:** A **contiguous block** of raw C-data. Strict typing, but lightning-fast for math.

**🧠 Memory Diagram:**
```text
PYTHON LIST: [10, 20, 30]
[ List Obj ] ──► [ Ptr ] ──► [ int: 10 ]
               ├─► [ Ptr ] ──► [ int: 20 ]
               └─► [ Ptr ] ──► [ int: 30 ]
(Overhead: Storing pointers + object metadata)

NUMPY ARRAY: np.array([10, 20, 30])
[ Array Obj ] ──► [ 10 | 20 | 30 ]
(Zero overhead, contiguous C-memory)
```

## 2. Lists vs. Arrays (NumPy/C++)

| Feature | Traditional Arrays | Python Lists |
| :--- | :--- | :--- |
| **Data Types** | **Homogeneous:** Must contain the same type. | **Heterogeneous:** Can mix ints, strings, floats, etc. |
| **Memory** | **Contiguous:** Stored sequentially in memory. | **References:** Stores pointers to scattered memory locations. |
| **Size** | **Fixed:** Capacity declared at creation. | **Dynamic:** Automatically resizes as needed. |
| **Performance**| Extremely fast for mathematical operations. | Slower due to dynamic typing and memory overhead. |

## 3. Characteristics of a List

1. **Ordered:** Items maintain their specific index position.
2. **Mutable:** You can change, add, or delete items on the fly.
3. **Allows Duplicates:** `[1, 1, 1]` is perfectly valid.
4. **Heterogeneous:** `[1, "Two", 3.0, True]` is valid.

---


````python
# Demonstrating Mutability and Heterogeneity
my_list = [10, "Hello", 3.14, True]
print("Original:", my_list)
````

````python
# Lists are mutable! We can change index 1
my_list[1] = "World"
print("Mutated:", my_list)
````

## 4. Editing and Deleting Items

Because lists are mutable, Python provides powerful tools to manage their contents.

* **Editing:** Use indexing (`L[0] = 5`) or slicing (`L[1:3] = [10, 20]`).
* **Deleting:**
  * `del L[index]`: Deletes item at a specific index.
  * `remove(value)`: Searches for a value and removes its *first* occurrence.
  * `pop(index)`: Removes and *returns* the item. If no index is provided, pops the last item.
  * `clear()`: Wipes the entire list, leaving `[]`.


````python
L = ["A", "B", "C", "D", "E"]

# Remove by value
L.remove("C")
print("After remove:", L)
````

````python
# Remove and capture by index
captured = L.pop() # Removes "E"
print("After pop:", L)
print("Captured item:", captured)
````

````python
# Delete by slice
del L[0:2]
print("After slice delete:", L)
````

## 5. Adding Items and List Methods

* **`append(item)`:** Adds a single item to the very end of the list.
* **`extend(iterable)`:** Adds multiple items (from another list or sequence) to the end.
* **`insert(index, item)`:** Injects an item at a specific index, shifting everything else right.
* **`sort()`:** Sorts the list in ascending order *in-place* (modifies original).
* **`reverse()`:** Reverses the list *in-place*.


````python
nums = [5, 2, 9, 1]

nums.append(10)
print("Append:", nums)
````

````python
nums.extend([20, 30])
print("Extend:", nums)
````

````python
nums.insert(2, 99) # Insert 99 at index 2
print("Insert:", nums)
````

````python
nums.sort()
print("Sorted:", nums)
````

## 6. List Comprehension

List comprehension provides a concise, "Pythonic" way to generate new lists. It combines a `for` loop and an optional `if` statement into a single line.

**Syntax:** `[expression for item in iterable if condition]`


````python
# Traditional approach:
squares_traditional = []
for x in range(1, 6):
    squares_traditional.append(x ** 2)

print(squares_traditional)
````

````python
# List Comprehension approach:
squares_comp = [x ** 2 for x in range(1, 6)]

print("Squares:", squares_comp)
````

````python
# Comprehension with a condition (Only even numbers)
even_squares = [x ** 2 for x in range(1, 11) if x % 2 == 0]

print("Even Squares:", even_squares)
````

````python
# Extracting first letters
names = ["Anuj", "Rahul", "Priya"]
initials = [name[0] for name in names]
print("Initials:", initials)
````

### ⚠️ Common Mistake (The Shallow Copy Trap)
```python
dataset_A = [1, 2, 3]
dataset_B = dataset_A      # ❌ Just creates a second name tag!
dataset_B.append(4)

print(dataset_A)           # [1, 2, 3, 4] <- dataset_A is also modified!
```
**Fix:** Use `dataset_B = dataset_A.copy()` or `dataset_B = dataset_A[:]`.

### 🎯 Interview Question
**Q: What is the Time Complexity of `list.append()` vs `list.insert(0, val)`?**
**Answer:** 
* `append()` is **O(1)** (Amortized). It adds to the end.
* `insert(0, val)` is **O(N)**. Python must shift every single existing element in memory one step to the right. *Never insert at the beginning of a massive list!*


## 7. Traversing & The `zip()` Function

### Traversing a List
Traversing means looping through the elements of the list one by one.

1. **Item-wise Traversal:** `for item in list:` loops directly over the values. This is preferred because it is clean and highly readable.
2. **Index-wise Traversal:** `for i in range(len(list)):` loops over a generated range of index numbers (`range(len())`). This is necessary when you need to know the exact position of the item or when you need to modify the list during the loop.

*Pro Tip:* Use the built-in `enumerate()` function if you need **both** the index and the item simultaneously. It is the gold standard for Python iteration.

````python
L = ["Data", "Science", "Python"]

# 1. Item-wise Traversal
print("--- Item-wise ---")
for item in L:
    print(item)
````

````python
# 2. Index-wise Traversal
print("\n--- Index-wise ---")
for i in range(len(L)):
    print(f"Index {i} holds: {L[i]}")
````

````python
# 3. The Enumerate Way (Best of both worlds)
print("\n--- Using Enumerate ---")
for index, item in enumerate(L):
    print(f"Position {index} -> {item}")
````

### The `zip()` Function

The `zip()` function is incredibly useful for parallel iteration. It takes two or more iterables (like lists) and "zips" them together, pairing the 1st item of List A with the 1st item of List B, the 2nd with the 2nd, and so on. It returns an iterator of tuples.

*(Note: If the lists are of different lengths, `zip()` stops at the end of the shortest list).*

````python
# The power of zip()
names = ["Anuj", "Ravi", "Sneha"]
scores = [95, 88, 92]
subjects = ["Math", "Physics", "Chemistry"]

# 1. Zipping three lists together simultaneously
for name, score, subject in zip(names, scores, subjects):
    print(f"{name} scored {score} in {subject}.")
````

````python
# 2. Creating a Dictionary from two zipped lists
name_score_dict = dict(zip(names, scores))
print("\nCreated Dictionary:", name_score_dict)
````

````python
# 3. Unzipping (Separating them back using the unpack operator *)
zipped_list = list(zip(names, scores))
unzipped_names, unzipped_scores = zip(*zipped_list)
print("\nUnzipped Names:", unzipped_names)
````

### ⚠️ Common Mistake
```python
# Modifying a list while iterating over it
numbers = [1, 2, 2, 3]
for n in numbers:
    if n == 2:
        numbers.remove(n) # ❌ Skips elements! Messes up the internal index counter.
```
**Fix:** Iterate over a copy `for n in numbers[:]:` or use List Comprehension.

## 8. Disadvantages of Lists
While Python lists are the ultimate "Swiss Army knife" for general programming, they have distinct drawbacks when scaling up to massive datasets in Data Science and Machine Learning.

1. **Memory Overhead:** Because lists allow heterogeneous data, they don't store raw bytes sequentially. Instead, they store a contiguous array of *pointers* (memory addresses) that point to actual objects scattered across RAM. This object metadata costs significantly more memory than a raw C-array.
2. **Execution Speed:** Because of dynamic typing, Python must perform "type-checking" on every single element during an operation (e.g., checking if the item is an int or a string before adding). This makes loops over millions of list items very slow.
3. **No Vectorized Math:** You cannot easily perform element-wise mathematical operations. For example, `[1, 2] * 2` results in `[1, 2, 1, 2]` (repetition), not `[2, 4]` (multiplication).
4. **Inefficient Insertions/Deletions in the Middle:** Inserting or deleting an item at the beginning or middle of a list (`insert(0, val)`) requires shifting all subsequent elements in memory, giving it an **O(n)** time complexity.
5. **Linear Search Time:** Checking if an item exists (`x in list`) takes **O(n)** time because it must scan the list sequentially from the beginning.

**The Solution:** * In Data Science, we solve math and speed disadvantages by replacing standard Python lists with **NumPy Arrays**, which enforce strict typing, use contiguous C-level memory, and support lightning-fast vectorized mathematics.
* For fast insertions/deletions at both ends, we use **`collections.deque`**.
* For fast O(1) lookups, we use **Sets** or **Dictionaries**.

````python
# Demonstrating the speed difference (conceptual)
# Python List (slow for math)
py_list = [1, 2, 3]
print("List multiplication:", py_list * 2)  # [1, 2, 3, 1, 2, 3] - NOT element-wise!
````

````python
# NumPy Array (fast for math)
import numpy as np
np_array = np.array([1, 2, 3])
print("NumPy multiplication:", np_array * 2)  # [2, 4, 6] - Element-wise!
````

## Tuples (The Immutable Sequences)

### 1. What are Tuples? (DS Context)
* **Definition:** An ordered, **immutable** collection of items. 
* **DS Context:** While lists are used for homogeneous data of variable length (like a column of ages), tuples are used for **heterogeneous data of fixed length** (like a single row in a database: `("P001", "John", 25)`).

### 2. Creating a Tuple & Special Syntax


````python
# Standard creation
patient_record = ("P001", "John Doe", 25)

# ⚠️ SPECIAL SYNTAX: Single-element tuple
# Without the comma, Python thinks it's just math parentheses!
t1 = (10)       # ❌ Type is int
t2 = (10,)      # ✅ Type is tuple (The comma is mandatory)

# Without parentheses (Tuple Packing)
t3 = 1, 2, 3    # ✅ Creates a tuple automatically

print(t1, type(t1), t2, type(t2), t3, type(t3), sep="\n")
````

### 3. Accessing, Editing, Adding, Deleting
Because tuples are **immutable**, you cannot change them in place.


````python
data = (10, 20, 30, 40)

# Accessing (Exactly like lists)
print(data[0])      # 10
print(data[-1])     # 40
print(data[1:3])    # (20, 30)

# Editing/Adding/Deleting? 
# data[0] = 99      # ❌ TypeError: 'tuple' object does not support item assignment
# data.append(50)   # ❌ AttributeError
````

````python
# Workaround 1: Reassignment via concatenation
data = data + (50,)  # Creates a brand new tuple in memory
print("After adding 50:", data)
````

````python
# Workaround 2: The "Mutable Inner Object" Exception
# If a tuple contains a mutable object (like a list), you CAN mutate the inner object!
mixed_tuple = (1, 2, [3, 4])
mixed_tuple[2][0] = 99  # ✅ Valid! The tuple still points to the same list, the list just changed.
print(mixed_tuple)      # (1, 2, [99, 4])
````

### 4. Operations & Functions

````python
t1 = (1, 2, 3)
t2 = (1, 2, 3)

# Operations
print(t1 == t2)     # True (Value comparison)
````

````python
print(t1 * 2)       # (1, 2, 3, 1, 2, 3)
````

````python
print(2 in t1)      # True
````

````python
# Functions (Very limited compared to lists)
print(len(t1))      # 3
````

````python
print(t1.count(2))  # 1 (Counts occurrences)
````

````python
print(t1.index(3))  # 2 (Finds first index)
````

### 5. Lists vs. Tuples (The 7 Pillars)

| Feature | Lists `[]` | Tuples `()` |
| :--- | :--- | :--- |
| **1. Syntax** | Square brackets `[]` | Parentheses `()` |
| **2. Mutability** | **Mutable** (Can change) | **Immutable** (Cannot change) |
| **3. Speed** | Slower to iterate and create. | **Faster** (Python optimizes immutable objects). |
| **4. Memory** | **Over-allocated.** Python allocates extra memory to make `append()` fast. | **Exact allocation.** No extra memory. Highly memory-efficient. |
| **5. Built-in Funcs**| Many (`append`, `pop`, `sort`, etc.) | Very few (`count`, `index`). |
| **6. Error Prone** | High. Accidental modification can corrupt data. | **Zero.** Guarantees data integrity. |
| **7. Usability** | Homogeneous data, variable length (Iterables). | Heterogeneous data, fixed length (Records/Structures). |

### 6. Why Use Tuples?
1. **Data Integrity:** When you want to ensure data is never accidentally modified (e.g., configuration settings, database rows).
2. **Dictionary Keys:** Because tuples are immutable (and hashable), they can be used as dictionary keys. Lists cannot.
3. **Memory & Speed:** If you have millions of constant records, tuples save massive amounts of RAM and process faster.
4. **Tuple Unpacking:** A powerful Pythonic feature.


````python
# Tuple Unpacking (Highly used in ML for returning multiple values)
def get_patient_stats():
    return 120, 80, 98.6  # Returns a tuple

systolic, diastolic, temp = get_patient_stats() 
````

### 🎯 Interview Question
**Q: Since tuples are immutable, how does Python achieve faster creation speeds than lists?**
**Answer:** Python uses a **free-list** memory optimization for tuples. Because tuples are immutable and fixed-size, Python can safely reuse the exact same memory addresses for small tuples, avoiding the overhead of dynamic memory allocation that lists require.


##  Sets (The Mathematical Collections)

### 1. Characteristics of Sets
* **Unordered:** Items have no specific index. (Note: Python 3.7+ maintains insertion order for *dicts*, but sets remain conceptually unordered).
* **Mutable:** You can add/remove items, but you cannot change the item itself.
* **No Duplicates:** Automatically filters out duplicates.
* **Hashable Elements Only:** Can only contain immutable data types (ints, strings, tuples). **Cannot contain lists or dicts.**

### 2. Creating a Set & The Empty Set Gotcha


````python
# Standard creation
unique_ids = {"P001", "P002", "P003"}
````

````python
# ⚠️ THE EMPTY SET GOTCHA
empty_dict = {}       # ❌ Creates an empty DICTIONARY!
empty_set = set()     # ✅ Creates an empty SET.
````

````python
# Creating from a list (Great for removing duplicates!)
ages = [25, 22, 25, 30, 22]
unique_ages = set(ages)  # {25, 22, 30}
print("Unique ages and ages type:", type(unique_ages), type(ages))
````

### 3. Accessing, Editing, Adding, Deleting
Because sets are **unordered**, you cannot use indexing (`s[0]` is illegal).


````python
s = {"apple", "banana", "cherry"}

# Accessing: Must iterate or convert to list
print("apple" in s)  # True (Membership testing is blazing fast O(1))

# Editing: Must remove old, add new
# s[0] = "mango"     # ❌ TypeError
````

````python
# Adding
s.add("mango")           # Adds one item
print("After adding mango:", s)
````

````python
s.update(["kiwi", "pear"]) # Adds multiple items
print("After updating with multiple items:", s)
````

````python
# Deleting
s.remove("banana")       # ❌ Throws KeyError if item not found!
print("After removing banana:", s)
````

````python
s.discard("banana")      # ✅ Safe. Does nothing if item not found.
print("After discarding banana:", s)
````

````python
s.pop()                  # Removes and returns a random item (since it's unordered)
print("After popping an item:", s)
````

````python
s.clear()                # Empties the set
print("After clearing the set:", s)
````

### 4. Set Operations (Venn Diagram Logic)
Crucial for data science when comparing datasets (e.g., finding overlapping patients between two clinics).


````python
clinic_A = {"P001", "P002", "P003"}
clinic_B = {"P002", "P003", "P004"}

# 1. Union (|): All unique patients from both
print(clinic_A | clinic_B)  # {'P001', 'P002', 'P003', 'P004'}
````

````python
# 2. Intersection (&): Patients in BOTH clinics
print(clinic_A & clinic_B)  # {'P002', 'P003'}
````

````python
# 3. Difference (-): Patients in A but NOT in B
print(clinic_A - clinic_B)  # {'P001'}
````

````python
# 4. Symmetric Difference (^): Patients in A or B, but NOT both
print(clinic_A ^ clinic_B)  # {'P001', 'P004'}
````

### 5. Frozenset
* **Definition:** An immutable version of a set. 
* **Why?** Standard sets are mutable, so they cannot be used as dictionary keys or elements inside another set. Frozensets solve this.


````python
fs = frozenset([1, 2, 3])
# fs.add(4)  # ❌ AttributeError

my_dict = {fs: "Valid Key"} # ✅ Works perfectly!
````

##  Dictionaries (The Hash Maps)

### 1. Characteristics
* **Mutable:** Can change, add, and remove key-value pairs.
* **Indexing has no meaning:** Accessed via **Keys**, not numeric indices.
* **Keys can't be duplicated:** If you assign a value to an existing key, it overwrites the old value.
* **Keys can't be mutable:** Keys must be hashable (strings, numbers, tuples). Values can be anything.
* **Order:** Since Python 3.7+, dictionaries **maintain insertion order**.
### 2. Creating a Dictionary


````python
# Standard creation
patient1 = {
    "id": "P001",
    "name": "John Doe",
    "age": 25
}
print("Patient 1 Record:", patient1)
````

````python
# Using dict() constructor
patient2 = dict(id="P002", name="Jane", age=30)
print("Patient 2 Record:", patient2)
````

### 3. Accessing Items (`[]` vs `.get()`)


````python
# Method 1: Bracket Notation
print(patient1["name"])   # "John Doe"
# print(patient["gender"]) # ❌ KeyError: Crashes the program!
````

````python
# Method 2: The .get() method (Highly Recommended for DS)
print(patient1.get("gender"))          # None (Doesn't crash)
print(patient2.get("gender", "Unknown")) # "Unknown" (Returns default value)
````

### 4. Editing & Adding Key-Value Pairs


````python
# Adding (If key doesn't exist, it creates it)
patient1["gender"] = "Male"
````

````python
# Editing (If key exists, it overwrites)
patient1["age"] = 26
````

````python
# Updating multiple keys at once
patient1.update({"age": 27, "city": "New York"})
````

### 5. Deleting vs. Removing Key-Value Pairs
* **Deleting (`del`, `clear`):** Destroys the data.
* **Removing (`pop`, `popitem`):** Deletes the data **AND returns it** so you can use it.


````python
# 1. del keyword
del patient1["city"]          # Deletes "city"
# del patient                # Deletes the entire dictionary from memory
````

````python
# 2. pop(key)
removed_age = patient1.pop("age") 
print(removed_age)           # 27 (Returns the value)
````

````python
# 3. popitem()
last_item = patient1.popitem() # Removes and returns the LAST inserted (key, value) tuple
print(last_item)             # ('name', 'John Doe')
````

````python
# 4. clear()
patient1.clear()              # Empties the dict: {}
print("After clearing patient1:", patient1)
````


### 6. Dictionary Operations & Functions

````python
data = {"a": 1, "b": 2}

# Operations
print("a" in data)       # True (Checks KEYS only, not values)
print(len(data))         # 2
````

````python
# Python 3.9+ Merge Operator (|)
d1 = {"a": 1}
d2 = {"b": 2}
merged = d1 | d2         # {'a': 1, 'b': 2}

print("Merged Dictionary:", merged)
````

````python
# Functions / Views
print(data.keys())       # dict_keys(['a', 'b'])
print(data.values())     # dict_values([1, 2])
print(data.items())      # dict_items([('a', 1), ('b', 2)]) -> Crucial for looping!
````

### 7. Dictionary Comprehension
The Pythonic way to transform data (e.g., swapping keys and values).


````python
original = {"a": 1, "b": 2, "c": 3}

# Square the values
squared = {k: v**2 for k, v in original.items()}
# {'a': 1, 'b': 4, 'c': 9}
````

````python
# Swap keys and values
swapped = {v: k for k, v in original.items()}
# {1: 'a', 2: 'b', 3: 'c'}
````

### 🧠 Memory Diagram (How Dicts Work Under the Hood)
Dictionaries use a **Hash Table**.
```text
1. Hash Function: hash("name") -> 1402
2. Index Calculation: 1402 % array_size -> 2
3. Storage: 
   [ Index 0 ] [ Index 1 ] [ Index 2: ("name", "John") ] ...
```
*This is why lookups are O(1). Python does the math to jump directly to the memory address, no searching required!*

### ⚠️ Common Mistake
```python
# Using a mutable type as a key
my_dict = {[1, 2]: "value"}  # ❌ TypeError: unhashable type: 'list'
```
**Fix:** Use a tuple instead: `my_dict = {(1, 2): "value"}`

### 🎯 Interview Question
**Q: What happens if two different keys hash to the same index in a dictionary? (Hash Collision)**
**Answer:** Python handles collisions using **Open Addressing** (specifically, probing). If the calculated index is already occupied, Python checks the next available slot in the array until it finds an empty one. Because of this, dictionary performance degrades if the table gets too full, which is why Python resizes the hash table dynamically.
