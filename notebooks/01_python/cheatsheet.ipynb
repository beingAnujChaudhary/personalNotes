# 📚 MODULE 1: Python Memory & Variables

### 1. Definition
Variables in Python are not "buckets" that store data; they are **references** (name tags) that point to objects located in memory.

### 2. Why do we need it?
Understanding this is crucial for avoiding shallow-copy bugs, understanding function argument passing (pass-by-object-reference), and optimizing memory in large ML pipelines.

### 3. Characteristics
*   **Dynamic Typing:** No type declaration needed; inferred at runtime.
*   **Dynamic Binding:** A variable can point to an `int`, then later point to a `str`.
*   **Garbage Collection:** Python uses reference counting. When an object's reference count drops to 0, it is destroyed.

### 4. Internal Working
Every object in Python has a header containing its **type**, **reference count**, and **value**. Variables are simply entries in a namespace (a dictionary) mapping a string name to a memory address (`id`).

### 5. Memory Diagram
```text
Code: a = [1, 2]
      b = a

Namespace (Variables)       Memory (Objects)
[a] ─────────────────┐
                     ├──► [ List Obj | ID: 101 | Ref: 2 | Val: 1, 2 ]
[b] ─────────────────┘
```

### 6. Syntax
```python
variable_name = value
```

### 7. Examples
```python
x = 10
print(id(x))  # Output: 140409392 (Memory address)
x = "Hello"
print(id(x))  # Output: 140409552 (Points to a completely new object)
```

### 8. Data Science Examples
When passing a massive Pandas DataFrame to a function, Python doesn't copy the data; it passes a reference. Modifying the DataFrame inside the function modifies the original dataset in memory, saving gigabytes of RAM.

### 9. Time Complexity
*   **Variable Assignment:** O(1)
*   **Type Checking:** O(1)

### 10. Common Mistakes
**The Aliasing Trap:**
```python
dataset_A = [1, 2, 3]
dataset_B = dataset_A      # ❌ Creates a second reference, NOT a copy!
dataset_B.append(4)
print(dataset_A)           # Output: [1, 2, 3, 4] <- Original is corrupted!
```
**Fix:** Use `dataset_B = dataset_A.copy()` or `dataset_B = dataset_A[:]`.

### 11. Interview Questions
**Q: What is the difference between `==` and `is`?**
**A:** `==` checks **value equality** (calls the `__eq__()` method). `is` checks **identity** (compares memory addresses via `id(a) == id(b)`).
**Q: Why use `if x is None:` instead of `if x == None:`?**
**A:** `None` is a singleton object. `is` is faster because it compares memory addresses directly, and it is safer because custom classes can override `__eq__()` to behave unpredictably.

2. Write a function to check if two variables point to the exact same list object.

# 📚 MODULE 2: Lists (The Dynamic Arrays)

### 1. Definition
An ordered, mutable, heterogeneous sequence of items.

### 2. Why do we need it?
To store collections of data where the size changes dynamically (e.g., appending new data points during streaming).

### 3. Characteristics
*   **Ordered:** Maintains insertion order.
*   **Mutable:** Can be changed in place.
*   **Allows Duplicates:** Yes.

### 4. Internal Working
A Python list is a **contiguous array of pointers**. Each pointer points to a separate Python object elsewhere in memory. To make `append()` fast, Python **over-allocates** memory (allocates more space than currently needed) so it doesn't have to resize the array on every single append.

### 5. Memory Diagram
```text
PYTHON LIST: [10, 20]
[ List Obj | Over-allocated Space ]
   │
   ├─► [ Ptr ] ──► [ Int Obj: 10 ]
   ├─► [ Ptr ] ──► [ Int Obj: 20 ]
   └─► [ Ptr ] ──► [ Empty/Reserved Memory ]
```

### 6. Syntax
```python
my_list = [1, 2, 3]
```

### 7. Examples
```python
nums = [10, 20, 30]
nums.append(40)       # Output: [10, 20, 30, 40]
nums.insert(0, 5)     # Output: [5, 10, 20, 30, 40]
print(nums[1])        # Output: 10
```

### 8. Data Science Examples
Storing a single feature column (e.g., patient ages) before converting it to a NumPy array for vectorized matrix multiplication.

### 9. Time Complexity
*   **Index Access (`L[i]`):** **O(1)** (Direct pointer math)
*   **Append (`L.append(x)`):** **O(1)** (Amortized, due to over-allocation)
*   **Insert (`L.insert(i, x)`):** **O(N)** (Must shift all subsequent pointers)
*   **Search (`x in L`):** **O(N)** (Linear scan)

### 10. Common Mistakes
**Modifying while iterating:**
```python
nums = [1, 2, 2, 3]
for n in nums:
    if n == 2: nums.remove(n) # ❌ Skips elements! Internal index shifts.
```
**Fix:** Iterate over a copy `for n in nums[:]:` or use list comprehension.

### 11. Interview Questions
**Q: Why is `list.insert(0, val)` O(N) but `list.append(val)` O(1)?**
**A:** `append` adds to the end of the pre-allocated memory block. `insert(0)` requires shifting every single existing pointer one position to the right to make room at index 0.
**Q: Python List vs. NumPy Array?**
**A:** Lists store scattered pointers to heterogeneous objects (slow for math). NumPy stores contiguous, homogeneous raw C-data (fast, vectorized math).

2. Remove duplicates from a list while preserving the original order.

# 📚 MODULE 3: Tuples (The Immutable Records)

### 1. Definition
An ordered, **immutable** sequence of items.

### 2. Why do we need it?
To guarantee data integrity (prevent accidental modification), save memory, and use sequences as Dictionary keys.

### 3. Characteristics
*   **Immutable:** Cannot be changed after creation.
*   **Ordered:** Maintains insertion order.
*   **Hashable:** (Only if all internal elements are also hashable).

### 4. Internal Working
Like lists, tuples are arrays of pointers. However, because they are immutable, Python **does not over-allocate** memory; it allocates the exact size needed. Furthermore, Python caches small tuples in a "free-list" to make creation blazingly fast.

### 5. Memory Diagram
```text
TUPLE: (10, 20)
[ Tuple Obj | Exact Size | No Over-allocation ]
   │
   ├─► [ Ptr ] ──► [ Int Obj: 10 ]
   └─► [ Ptr ] ──► [ Int Obj: 20 ]
```

### 6. Syntax
```python
my_tuple = (1, 2, 3)
single_tuple = (1,)  # Comma is mandatory!
```

### 7. Examples
```python
coords = (10, 20)
# coords[0] = 15  # ❌ TypeError: Tuples are immutable
x, y = coords      # Tuple unpacking
```

### 8. Data Science Examples
Returning multiple metrics from a model evaluation function: `return accuracy, precision, recall`. Representing a single fixed row from a SQL database.

### 9. Time Complexity
*   **Index Access:** **O(1)**
*   **Creation:** Faster than lists (due to free-list caching).
*   **Memory:** Smaller than lists (no over-allocation).

### 10. Common Mistakes
**The Single-Element Trap:**
```python
x = (10)
print(type(x))  # ❌ Output: <class 'int'> (Math parentheses)
y = (10,)
print(type(y))  # ✅ Output: <class 'tuple'>
```

### 11. Interview Questions
**Q: Why are tuples faster and use less memory than lists?**
**A:** Tuples have exact memory allocation (no over-allocation overhead) and Python reuses memory addresses for small tuples via a free-list, avoiding the overhead of dynamic memory allocation.
**Q: Can a tuple contain a list? Is it hashable?**
**A:** Yes, a tuple can contain a list `(1, 2, [3, 4])`. However, it becomes **unhashable** because its contents can mutate, violating the requirement for a constant hash value.

### 12. Practice Problems
1. Swap two variables `a, b = 10, 20` without a temporary variable.
2. Create a dictionary where the keys are 2D coordinates (tuples).

# 📚 MODULE 4: Sets (The Hash Tables)

### 1. Definition
An unordered collection of unique, hashable items.

### 2. Why do we need it?
For blazing-fast membership testing (`O(1)`) and mathematical set operations (Union, Intersection).

### 3. Characteristics
*   **Unordered:** No indexing.
*   **Mutable:** Can add/remove items.
*   **Unique:** Automatically discards duplicates.

### 4. Internal Working
Sets use a **Hash Table**. When you add an item, Python runs `hash(item)` to calculate an integer, then uses modulo arithmetic to find an index in an underlying array. Collisions are resolved via open addressing (probing).

### 5. Memory Diagram
```text
Item: "apple"
1. hash("apple") -> 840291
2. 840291 % array_size -> Index 5
3. Store pointer at Array[5]
```

### 6. Syntax
```python
my_set = {1, 2, 3}
empty_set = set()  # {} creates a dict!
```

### 7. Examples
```python
s = {1, 2, 2, 3}
print(s)            # Output: {1, 2, 3}
s.add(4)
print(2 in s)       # Output: True
```

### 8. Data Science Examples
Finding overlapping patient IDs between two hospital databases (`set_A & set_B`). Converting a massive list of categorical variables to a set to find unique classes before one-hot encoding.

### 9. Time Complexity
*   **Add/Remove:** **O(1)** average.
*   **Membership (`x in s`):** **O(1)** average.
*   **Worst Case:** O(N) if there are massive hash collisions (rare).

### 10. Common Mistakes
**Creating an empty set:**
```python
x = {}      # ❌ Creates an empty Dictionary!
y = set()   # ✅ Creates an empty Set.
```

### 11. Interview Questions
**Q: How do you find common elements between two lists of 10 million items efficiently?**
**A:** Do NOT use nested loops `O(N^2)`. Convert one list to a Set `O(N)`, then iterate through the second list checking `if item in set` `O(1)`. Total time: **O(N)**.
**Q: Why can't a list be a set element or dict key?**
**A:** Because lists are mutable. If a list changes, its hash value would change, breaking the hash table's ability to find the item.

### 12. Practice Problems
1. Find the symmetric difference between two sets.
2. Write a function to check if two strings are anagrams using sets.

# 📚 MODULE 5: Dictionaries (The Key-Value Maps)

### 1. Definition
An ordered (Python 3.7+), mutable collection of key-value pairs.

### 2. Why do we need it?
For fast lookups, structured data representation (JSON), and caching/memoization.

### 3. Characteristics
*   **Key-Value:** Accessed via keys, not numeric indices.
*   **Unique Keys:** Assigning to an existing key overwrites the value.
*   **Immutable Keys:** Keys must be hashable.

### 4. Internal Working
Since Python 3.6, dicts use a dual-array structure: a **sparse hash table** (for fast lookup) and a **dense array** (to preserve insertion order and save memory). `hash(key)` points to the sparse table, which points to the dense array containing the `(key, value)` pair.

### 5. Memory Diagram
```text
Key: "name"
1. hash("name") -> Sparse Index 2
2. Sparse[2] -> Points to Dense Array Index 0
3. Dense[0] -> ("name", "Anuj")
```

### 6. Syntax
```python
my_dict = {"key": "value"}
```

### 7. Examples
```python
d = {"a": 1, "b": 2}
print(d.get("c", 0))  # Output: 0 (Safe access)
d["c"] = 3            # Add/Update
```

### 8. Data Science Examples
Parsing JSON API responses. Counting token frequencies in NLP (`collections.Counter` is a dict subclass). Caching expensive model inference results.

### 9. Time Complexity
*   **Get/Set/Delete:** **O(1)** average.
*   **Keys/Values View:** O(1) to access the view object.

### 10. Common Mistakes
**The KeyError Crash:**
```python
d = {"a": 1}
val = d["b"]  # ❌ KeyError!
val = d.get("b", 0)  # ✅ Returns 0 safely.
```

### 11. Interview Questions
**Q: What happens if two keys hash to the same index? (Hash Collision)**
**A:** Python uses open addressing (probing). It checks the next available slot in the sparse array until it finds an empty one or the matching key.
**Q: How do you merge two dictionaries in Python 3.9+?**
**A:** Use the merge operator: `merged = d1 | d2`.

### 12. Practice Problems
1. Invert a dictionary (swap keys and values).
2. Group a list of dictionaries by a specific key (e.g., group patients by 'ward').

### 9. Set Operations Quick Reference
| Math Concept | Python Operator | Method | Example Result |

| **Symmetric Diff** (A or B, not both)| `A ^ B` | `A.symmetric_difference(B)`| Elements in A or B, but not both|

### 10. Mutability & Hashability Rules
* **Rule:** To be a **Dictionary Key** or a **Set Element**, an object MUST be **Hashable**.
* **Hashable means:** The object must be **Immutable** and have a `__hash__()` method.
* **Hashable Types:** `int`, `float`, `str`, `tuple` (if it only contains hashable items), `frozenset`.
* **Unhashable Types:** `list`, `set`, `dict`.
* **Gotcha:** A tuple containing a list `(1, 2, [3, 4])` is **unhashable**.
