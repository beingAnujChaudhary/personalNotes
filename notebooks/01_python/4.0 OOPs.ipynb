# 🚀 Object Oriented Programming

## 1. OOP Theory: What, Why, and How?

### What is OOP?
Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around **data (objects)** rather than functions and logic. It maps real-world entities into code to make it more modular, reusable, and scalable.
* **Procedural Programming** is like a recipe: a list of instructions that operate on data.
* **OOP** is like creating a living ecosystem: you create entities (objects) that have their own data and know how to behave.

### What is an Object?
An **Object** is a real-world entity that has:
1. **State / Data:** Properties or attributes (e.g., color, brand, weight).
2. **Behavior:** Actions it can perform (e.g., move, break, start).
In Python, *everything* is an object (integers, strings, lists, etc.).

Example can be a **car**. It has both properties/attributes (color, brand and price) and behaviours/methods (start, stop, accelerate).

### Class & Object
* **Class:** A blueprint, template, or prototype for creating objects. It defines the structure, data, and behavior that its objects will have. (e.g., `Car` blueprint).
* **Object:** An instance of a class.  It is a concrete entity that has:
> - **State/Attributes**: Data stored in variables (fields)
> - **Behavior/Methods**: Functions that operate on the data
> - **Identity**: Unique memory location.

### 🎯 Object Characteristics
| Property | Description | Example |
|----------|-------------|---------|
| **State** | Current data values | balance = 5000 |
| **Behavior** | What it can do | withdraw(), deposit() |
| **Identity** | Unique reference | Memory address: 0x7f8b... |

It is the actual, physical manifestation of the blueprint in memory. (e.g., a specific red Honda Civic Car).

### 🏛️ Four Pillars of OOP
| Pillar | Description | Real-World Example |
|--------|-------------|-------------------|
| **Encapsulation** | Hiding internal details | ATM machine hides internal cash counting |
| **Inheritance** | Child inherits from parent | SportsCar inherits from Car |
| **Polymorphism** | Same interface, different behavior | `+` works for int, str, list differently |
| **Abstraction** | Showing only essential features | TV remote hides complex circuitry |

### 🎤 Interview Questions
1. **What is the difference between a class and an object?**
>   *Answer:* A class is a logical blueprint that takes up no memory, while an object is a physical instance of that class that consumes memory space.
2. **Why do we need OOP?**
>   *Answer:* OOP provides code reusability (Inheritance), data security (Encapsulation), reduces complexity (Abstraction), and allows flexibility (Polymorphism).
3. **Is Python Object-Oriented?**
 >  *Answer:* Yes. In Python everything is an object.
4. **What is OOP and why is it better than procedural programming?**
>   *Answer:* OOP organizes code around objects that bundle data and methods together. It's better because it provides better modularity, code reuse (inheritance), data security (encapsulation), and models real-world problems more naturally.
5. **What is the difference between a class and an object?**
>   *Answer:* A class is a blueprint/template that defines attributes and methods. An object is an instance of a class with actual values. Class exists once; objects can be many.
6. **What is OOP and why is it better than procedural programming?**
>   *Answer:* Instance variables (defined with `self.`) are unique to each object. Class variables (defined outside `__init__`) are shared among all objects of that class.


````python
# Basic Class and Object Example

class Dog:
    #Class Attribute (shared by all instances/objects)
    species = "Canine"

    # Constructor (Initialises object state/attributes)
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Instance Method (behaviour of the object)
    def bark(self):
        return f"{self.name} says Woof!"

# Creating an object (instance of the class)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

print(dog1.name, dog2.name)  # Output: Buddy Max
print(dog1.age, dog2.age)   # Output: 3 5
print(dog1.species, dog2.species)  # Output: Canine Canine
print(dog1.bark())  # Output: Buddy says Woof!
print(dog2.bark())  # Output: Max says Woof!
````

````python
# Class vs Object Comparison
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

# Class: Blueprint
print(Car)  # <class '__main__.Car'>
````

````python
# Objects: Actual instances
car1 = Car("Toyota", "Camry")
car2 = Car("Honda", "Civic")

print(f"Car1: {car1.brand} {car1.model}")  # Toyota Camry
print(f"Car2: {car2.brand} {car2.model}")  # Honda Civic
````

````python
# Prove they are different objects
print(car1 is car2)  # False
````

### What is Constructor?

A **constructor** is a special method (`__init__`) that executes automatically when an object is created and is used to initialize object state/data.

### Why Constructor?

The constructor is used to write the configuration related to the code/task, which needs to be initialized without the user's consent and before the app runs. Such as:
1. Database Connection
2. API Configuration
3. User Setup
4. Default Values

**Question: ** **The God Analogy:**

**Answer:** The constructor is used to write the configuration related to the code/task, which needs to be initialized without the user's consent and before the app runs.

### 🔑 Key Points about Constructor
1. Named `__init__` (double underscore init)
2. Called automatically when object is created
3. Used to initialize object attributes
4. Cannot return any value (returns None)
5. `self` parameter is mandatory (refers to current object)

## 2. Building a Banking Application (ATM Class)

To understand OOP, let's build an `Atm` class. 

### Class Diagram / Structure
Before coding, software engineers design a **Class Diagram** (Unified Modeling Language, i.e., Unified Modeling Language). It contains three parts:
1. **Class Name:** `ATM`
2. **Data/Attributes:** `pin`, `balance`
3. **Methods/Behaviors:** `check_balance()`, `withdraw()`, `deposit()`

*Note on Visibility:* 
* `+` means **Public** (Visible outside the class).
* `-` means **Private** (Not visible outside the class, achieved using `__` in Python).
```
+-------------------------+
|          ATM            |  <-- Class Name
+-------------------------+
| - __pin                 |  <-- Private Attributes (-)
| - __balance             |
+-------------------------+
| + check_balance(pin)    |  <-- Public Methods (+)
| + withdraw(pin, amount) |
| - __authenticate(pin)   |  <-- Private Methods (-)
+-------------------------+
```

````python
class Atm:
    # 1. Constructor (Special Function)
    def __init__(self, pin, balance=0):
        # Attributes
        self.pin = pin
        self.balance = balance
        
        # Call the menu upon object creation
        self.menu()
        
    def menu(self):
        user_input = input("""
        Welcome to the ATM!
        1. Press 1 to create PIN
        2. Press 2 to change PIN 
        3. Press 3 to deposit
        4. Press 4 to withdraw
        5. Press 5 to check balance
        6. Press 6 to exit
        """)
        
        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.deposit()
        elif user_input == '4':
            self.withdraw()
        elif user_input == '5': 
            self.check_balance()
        elif user_input == '6':
            print("Thank you for using the ATM. Goodbye!")
        else:
            print("Invalid option. Please try again.")

    # 2. ATM Functionalities
    def create_pin(self):
        self.pin = input("Enter your new PIN: ")
        print("PIN set successfully!")

    def change_pin(self):
        temp = input("Enter your current PIN: ")
        if temp == self.pin:
            self.pin = input("Enter your new PIN: ")
            print("PIN changed successfully!", self.pin)
        else:
            print("Invalid PIN")

    def deposit(self):
        temp = input("Enter your PIN: ")
        if temp == self.pin:
            amount = int(input("Enter amount to deposit: "))
            self.balance += amount
            print("Deposit successful!")
        else:
            print("Invalid PIN")

    def withdraw(self):
        temp = input("Enter your PIN: ")
        if temp == self.pin:
            amount = int(input("Enter amount to withdraw: "))
            if amount <= self.balance:
                self.balance -= amount
                print("Withdrawal successful!")
            else:
                print("Insufficient balance!")
        else:
            print("Invalid PIN")

    def check_balance(self):
        temp = input("Enter your PIN: ")
        if temp == self.pin:
            print(f"Your current balance is: ₹{self.balance}")
        else:
            print("Invalid PIN")

# Creating an object of the ATM class
sbi = Atm(str(546), 1000) 
````

## 3. The Mechanics of OOP: Functions, Self, and Constructors

### Methods vs. Functions
* **Function:** An independent block of code. You call it directly (e.g., `len(my_list)`).
* **Method:** A function defined *inside* a class. It can only be called by an object of that class (e.g., `my_list.append()`).

| Feature | Function | Method |
| :--- | :--- | :--- |
| **Definition** | A standalone block of code. | A function associated with an Object/Class (i.e., inside Object/Class). |
| **Belongs to** | Module/Global scope | Class/Object |
| **Invocation** | Called by name: `my_func()` | Called on an object: `obj.method()` |
| **First parameter** | No `self` needed | Must have `self` |
| **Called on** | Directly: `func()` | Object: `obj.method()` |
| **Data Access** | Only operates on data passed as arguments. | Automatically receives the object (`self`) and can access/modify its internal state, i.e., Object's attributes via self |
| **Dependency** | Independent. | Dependent on the object's state. |

### 🎤 Interview Questions
1. **Is every method a function?**
>  *Answer:* Yes
2. **Is every function a method?**
>  *Answer:* No
2. **What's the difference between a method and a function in Python**
>  *Answer:* A method is a function defined inside a class and takes `self` as the first parameter, allowing access to the object's attributes. A function is standalone and doesn't belong to any class.



### The Concept of `self` and The Golden Rule
**What is `self`?** 
**`self`** represents the **current instance** of the class. It's how an object refers to itself - like saying "me" or "myself" in English.

### 🔑 Key Points
1. `self` is NOT a Python keyword - it's a **convention** (you could use any name, but DON'T)
2. It must be the **first parameter** in instance methods
3. It refers to the **current object** being manipulated
4. Through `self`, you access the object's attributes and methods

### 🌍 Real-World Analogy

````python
class Student:

    def __init__(self,fname,lname):

        self.fname = fname
        self.lname = lname
        # Prove that 'self' is just the object itself!
        print(f"Memory address of self inside class: {id(self)}")
        
s1 = Student("Anuj", "Chaudhary")
# Student.__init__(s1,"Anuj", "Chaudhary") Alternative way to call the constructor

print(s1.fname, s1.lname)
# Python automatically passes object reference.

print(f"Memory address of object outside class: {id(s1)}")
# Notice that both memory addresses are exactly the same! 
# 'self' is simply 's1' wearing a disguise inside the class.
````

````python
### 🔬 How Self Works Internally

class Demo:
    def __init__(self, value):
        self.value = value
    
    def display(self):
        print(f"Value: {self.value}")
        print(f"self is: {self}")
        print(f"self type: {type(self)}")

obj = Demo(42)

# These two calls are identical!

obj.display()
Demo.display(obj)  # Python internally converts obj.method() to Class.method(obj)
````

````python
# Proof:
print(f"obj is: {obj}")
# The object and self point to the same memory location
````

### 🎤 Interview Questions
1. **Is `self` a keyword in Python?**
>   *Answer:* No. It is just a strong convention. You can name it `this`, `me`, or `potato`, but using `self` is standard Python practice.

2. **What is the difference between a method and a function?**
>   *Answer:* A method is bound to an object (defined inside a class and takes `self` as the first parameter). A function is not bound to any object.

3. **Why is it (`self`) needed?** 
> *Answer:* If you have 100 ATM objects, self ensures that when check_balance() is called, Python updates the balance of the correct ATM, not the others.

4. **How does Python internally handle self?** 
> *Answer:* When you call `obj.method(arg)`, Python internally converts it to `Class.method(obj, arg)`. The object itself is passed as the first argument.

## The Golden Rule of OOP
Every instance method must have self as first parameter.
**"Tell, Don't Ask" (Encapsulation)**
An object should manage its own state. Never allow external code to directly manipulate an object's internal data. Instead, expose **methods** that allow the object to modify its own data safely.
That is, Only an object of a class can access its methods and attributes. When a method is called, Python secretly passes the object itself as the first argument to the method. That received object is caught by the parameter we call `self`.
> * *Bad:* `atm.balance = 1000000` (Bypasses security!)
> * *Good:* `atm.withdraw(pin, 100)` (The object checks the PIN and logic itself).

## 4. Magic Methods (Dunder Methods) & The Fraction Class

### What are Magic Methods?
Magic methods are special methods in Python that start and end with double underscores (`__`). They are not meant to be invoked directly by you, but the invocation happens internally from the class on a certain action (like adding objects, printing objects, etc.).

They allow custom objects to behave like built-in objects.

### 🔮 Why Magic Methods?
- Enable operator overloading (`+`, `-`, `*`, `/`)
- Customize object behavior (`__str__`, `__len__`, `__getitem__`)
- Make custom classes work with Python's built-in functions
- Provide hooks for object lifecycle

### 📋 Common Magic Methods

| Magic Method | Purpose | Called When |
|--------------|---------|-------------|
| `__init__(self)` | Constructor | Object creation |
| `__str__(self)` | String representation | `print(obj)`, `str(obj)` |
| `__repr__(self)` | Official representation | `repr(obj)`, debugging |
| `__len__(self)` | Length | `len(obj)` |
| `__add__(self, other)` | Addition | `obj1 + obj2` |
| `__sub__(self, other)` | Subtraction | `obj1 - obj2` |
| `__mul__(self, other)` | Multiplication | `obj1 * obj2` |
| `__truediv__(self, other)` | Division | `obj1 / obj2` |
| `__eq__(self, other)` | Equality | `obj1 == obj2` |
| `__lt__(self, other)` | Less than | `obj1 < obj2` |
| `__getitem__(self, key)` | Indexing | `obj[key]` |
| `__call__(self)` | Make callable | `obj()` |


Let's build a `Fraction` class (like 1/2, 3/4) to demonstrate Operator Overloading using Magic Methods.

````python
class Fraction:
    def __init__(self, numerator, denominator):
        self.num = numerator
        self.den = denominator
        
    # Magic method triggered when we try to print() the object
    def __str__(self):
        return f"{self.num}/{self.den}"
        
    # Magic method triggered by the '+' operator
    def __add__(self, other):
        new_num = (self.num * other.den) + (other.num * self.den)
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    # Magic method triggered by the '-' operator
    def __sub__(self, other):
        new_num = (self.num * other.den) - (other.num * self.den)
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    # Magic method triggered by the '*' operator
    def __mul__(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
        
    # Magic method triggered by the '/' operator
    def __truediv__(self, other):
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)

# Let's test the Magic Methods!
x = Fraction(1, 2) # Represents 1/2
y = Fraction(3, 4) # Represents 3/4

print(f"Fraction X: {x}") # Triggers __str__
print(f"Fraction Y: {y}")

print(f"Addition (x + y): {x + y}")       # Triggers __add__
print(f"Subtraction (x - y): {x - y}")    # Triggers __sub__
print(f"Multiplication (x * y): {x * y}") # Triggers __mul__
print(f"Division (x / y): {x / y}")       # Triggers __truediv__
````

### 🎤 Interview Questions
1. **What happens under the hood when you write `print(obj)`?**
>   *Answer:* Python looks for the `__str__` magic method inside the class. If it finds it, it executes it and prints the returned string. If not, it prints the memory address of the object.

2. **How does Python know how to add two custom objects together?**
>   *Answer:* Python relies on Operator Overloading via magic methods. When it sees `obj1 + obj2`, it secretly calls `obj1.__add__(obj2)`. If `__add__` is defined in the class, it works; otherwise, it throws a TypeError.

3. **Why use `str`?**
>   *Answer:* To provide a user-friendly string representation of an object.

4. **What is encapsulation?**
> **Answer:** Bundling data and methods within a class and restricting direct access. In Python, prefix with `__` for name mangling to indicate private attributes.

5. **What is the difference between class variable and instance variable?**
> **Answer:** Class variables are shared by all objects (defined outside `__init__`). Instance variables are unique to each object (defined with `self.` inside `__init__`).

6. **How does Python handle method calls internally?**
> **Answer:** When `obj.method(arg)` is called, Python converts it to `Class.method(obj, arg)`. The object is passed as the first argument automatically.

7. **Why can't Python have multiple constructors like Java?**
> **Answer:** Python doesn't support constructor overloading. Only the last `__init__` is valid. You can simulate it using default arguments or class methods as alternative constructors.

8. **What is the Golden Rule of OOP?**
> **Answer:** "Everything in Python is an object!" - int, str, list, dict, functions, and even classes themselves are objects. This is fundamental to Python's design.

9. **Explain `__str__` vs `__repr__`.**
> **Answer:** `__str__` provides user-friendly output for `print()`. `__repr__` provides unambiguous developer output for debugging. If `__str__` is absent, `__repr__` is used as fallback.

10. **What is name mangling in Python?**
> **Answer:** Attributes prefixed with `__` (double underscore) are name-mangled to `_ClassName__attribute`. This prevents accidental access but doesn't make them truly private.
