## 4. Core Data Types

Data types define the kind of value a variable holds, dictating what operations can be performed on it.

1. **Int (Integer):** Whole numbers. Python can handle massive integers out-of-the-box (e.g., `1000000`). If evaluating an exponent too large (like `1e309`), Python evaluates it to `inf` (infinity).
2. **Float:** Decimal numbers (e.g., `3.14`, `1.7e308`).
3. **Boolean (`bool`):** Represents truth values: `True` or `False`.
4. **String (`str`):** Text wrapped in quotes (`"Anuj"` or `'Anuj'`).
5. **Complex Num:** Real + Imaginary parts, denoted with `j` (e.g., `3 + 4j`).
6. **NoneType:** A special type representing `None` (the absence of a value).

You can check the type of any data using the `type()` function.


````python
# Checking data types
print(type("Data Science")) # <class 'str'>
````

````python
print(type(100))            # <class 'int'>
````

````python
print(type(3.14))           # <class 'float'>
````

````python
print(type(3 + 4j))         # <class 'complex'>
````

````python
# Scientific notation
massive_num = 1.7e308
print("Massive Float:", massive_num)
````

## 5. Variables: Dynamic Typing and Binding

A variable is a named container for storing data in your computer's memory.

* **Static vs. Dynamic Typing:** In languages like Java or C++, you must declare a variable's type (`int x = 5;`). Python uses **Dynamic Typing**—you do not declare the type. Python infers it at runtime.
* **Dynamic Binding:** A single variable can store different types of data at different points in your script's execution.
* **Multiple Assignment:** You can assign multiple variables on a single line to keep code concise.


````python
# Dynamic Binding
a = 5         # 'a' is born as an integer
print("a is:", type(a))
a = "Anuj"    # 'a' metamorphoses into a string
print("a is now:", type(a))
````

````python
# Multiple Assignment (different values)
x, y, z = 1, 2.5, "Three"
print(x, y, z)
````

````python
# Multiple Assignment (same value)
score1 = score2 = score3 = 100
print(score1, score2, score3)
````

**🧠 Memory Diagram:**
```text
Step 1: age = 25
[Variable 'age'] ──────► [ Memory Object: 25 (Type: int) ]

Step 2: age = "Twenty-Five"
[Variable 'age'] ──────► [ Memory Object: "Twenty-Five" (Type: str) ]
                         (The old '25' object is garbage-collected if nothing else points to it)
```

## 8. Type Conversion, Literals, and Booleans

* **Type Conversion (Casting):** Changing one data type to another.
    * *Implicit:* Python does it automatically behind the scenes (e.g., `int` + `float` = `float`).
    * *Explicit:* The programmer forces it using `int()`, `float()`, `str()`, or `bool()`.
* **Literals:** Raw, fixed data directly written into the code (e.g., `5`, `"Hello"`, `True`).
* **Booleans (Truthy/Falsy):** Python evaluates almost any data with content as `True` (Truthy).
  * **Falsy values:** `0`, `0.0`, empty strings `""`, empty lists `[]`, and `None`.


````python
# Explicit Type Conversion
num_str = "100"
# We cannot do math on a string, so we explicitly cast it:
num_int = int(num_str)
print("Math operation successful:", num_int + 50)
````

````python
# Truthy and Falsy
print("Is 10 True?", bool(10))        # True
print("Is -5 True?", bool(-5))        # True (Negative numbers are Truthy!)
print("Is 0 True?", bool(0))          # False (Falsy)
print("Is '' True?", bool(""))        # False (Empty string is Falsy)
````

````python
# The None Type
placeholder = None
if placeholder is None:
    print("This variable is completely empty.")
````

# 🧮 Lecture 2: Operators

Operators are special symbols that carry out arithmetic or logical computation.

| Category | Operators | Description | Example |
| :--- | :--- | :--- | :--- |
| **Arithmetic** | `+`, `-`, `*`, `/`, `//`, `%`, `**` | Math calculations. Note: `//` is floor division (drops decimals), `**` is exponent. | `5 ** 2` = 25 |
| **Relational** | `>`, `<`, `>=`, `<=`, `==`, `!=` | Compares values. Always returns a Boolean. | `5 == 5` is True |
| **Logical** | `and`, `or`, `not` | Combines conditional statements. | `True and False` is False |
| **Assignment** | `=`, `+=`, `-=`, `*=` | Assigns or updates a variable's value in place. | `x += 5` (same as `x = x + 5`) |
| **Membership** | `in`, `not in` | Checks if a value exists within a sequence (like a list or string). | `'a' in 'cat'` is True |
| **Identity** | `is`, `is not` | Checks if two variables point to the exact same object in memory. | `a is b` |


````python
# Arithmetic Operators
print("10 / 3 =", 10 / 3)      # Standard float division
print("10 // 3 =", 10 // 3)    # Floor division (drops the remainder)
print("10 % 3 =", 10 % 3)      # Modulus (returns ONLY the remainder)
````

````python
# Logical Operators
age = 25
has_license = True
can_drive = (age >= 18) and has_license
print("Can drive?", can_drive)
````

````python
# Membership Operators
vowels = "aeiou"
print("Is 'z' a vowel?", 'z' in vowels)
````

### Identity vs. Equality Operators
* `==` checks if **values** are the same.
* `is` checks if they are the exact same **object in memory**.

**🧠 Memory Diagram:**
```text
a = [1, 2]
b = [1, 2]
c = a

[a] ──► [List Obj 1] <── [c]
[b] ──► [List Obj 2]

a == b  # True (Values match)
a is b  # False (Different memory addresses)
a is c  # True (Exact same memory address)

### ⚠️ Common Mistake
```python
# Floating Point Math Error
print(0.1 + 0.2 == 0.3)  # ❌ Outputs False!
```
**Why?** Computers store floats in binary (IEEE 754). `0.1 + 0.2` is actually `0.30000000000000004`.
**Fix:** Use `math.isclose(0.1 + 0.2, 0.3)` or `round()`.

### 🎯 Interview Question
**Q: What is the difference between `/` and `//`?**
**Answer:** `/` is float division (always returns a float, e.g., `5 / 2 = 2.5`). `//` is floor division (drops the decimal, e.g., `5 // 2 = 2`).


# Control Flow & Modules

## 1. If-Else Statements

Control flow structures allow your program to make decisions and execute specific blocks of code only if certain conditions are met.

* `if condition:` Evaluated first.
* `elif condition:` Evaluated if the preceding `if` was False.
* `else:` The catch-all. Executes if absolutely all conditions above it are False.


````python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
````

## 2. Loops in Python

Loops execute a block of code repeatedly. They are essential for adhering to the DRY (Don't Repeat Yourself) principle.

### The `while` Loop
Repeats as long as a boolean condition remains `True`. Beware of infinite loops!

### The `for` Loop
Used to iterate sequentially over a collection (like a string, list, or a `range()`).

### Loop Control Statements
* **`break`:** Terminates the loop entirely, breaking out of it immediately.
* **`continue`:** Skips the rest of the *current* iteration and jumps back to the top of the loop for the next iteration.
* **`pass`:** A null statement. It does literally nothing. Used as a placeholder to prevent syntax errors when writing empty code blocks.


````python
# While Loop
countdown = 3
while countdown > 0:
    print(countdown)
    countdown -= 1
print("Blastoff!")
````

````python
# For loop using range(start, stop, step)
# stop is exclusive!
for i in range(1, 10, 2):
    print("Odd number:", i)
````

````python
# Break and Continue
for num in range(1, 6):
    if num == 3:
        continue  # Skips 3
    if num == 5:
        break     # Stops the loop entirely before printing 5
    print("Num:", num)
````

# Strings

## 1. What are Strings?

Strings are sequences of characters enclosed in single `''` or double `""` quotes. **Crucial Concept:** Strings are **immutable**. Once created in memory, you cannot alter a specific character within that string. You can only create a new string.

### Indexing and Slicing
* **Positive Indexing:** Left to right, starting at `0`.
* **Negative Indexing:** Right to left, starting at `-1` (the last character).
* **Slicing:** Extracting a chunk using `[start : stop : step]`. The `stop` index is always exclusive.


````python
text = "Python Programming"

# Indexing
print("First character:", text[0])    # 'P'
print("Last character:", text[-1])    # 'g'
````

````python
# Slicing
print("First 6 characters:", text[0:6])  # 'Python'
print("Every 2nd character:", text[::2]) # 'Pto rgamn'
print("Reversed string:", text[::-1])    # 'gnimmargorP nohtyP'
````

---

## 2. String Functions and Methods

Strings come with a massive arsenal of built-in methods. Because strings are immutable, these methods always return a **new** string rather than changing the original.

* **Length:** `len(str)`
* **Transformations:** `upper()`, `lower()`, `capitalize()`, `title()`
* **Searching:** `find()` (returns index, -1 if not found), `count()` (counts occurrences)
* **Validation:** `isalnum()`, `isalpha()`, `isdigit()`, `startswith()`, `endswith()`
* **Utility:**
  * `split(delimiter)`: Chops a string into a List of strings.
  * `join(list)`: Glues a List of strings back together.
  * `replace(old, new)`: Swaps out text.


````python
s = "  data science is amazing  "

# Cleaning and Formatting
clean_s = s.strip() # Removes leading/trailing whitespace
print(clean_s.title()) # Capitalizes first letter of each word
````

````python
# Splitting and Joining
csv_data = "apple,banana,cherry"
fruits_list = csv_data.split(",")
print("List:", fruits_list)
````

````python
joined_string = " | ".join(fruits_list)
print("Joined:", joined_string)
````

````python
# Replacing
print(joined_string.replace("apple", "mango"))
````

````python
raw_data = "  Patient: John Doe, Age: 30  "
````

````python
# 1. Strip whitespace
clean_data = raw_data.strip() 
print("Cleaned Data:", clean_data)
````

````python
# 2. Split by delimiter
parts = clean_data.split(",")  # ['Patient: John Doe', 'Age: 30']
print("Parts:", parts)
````

````python
# 3. Replace anomalies
fixed = parts[0].replace("Patient:", "").strip() # 'John Doe'
print("Fixed Name:", fixed)
````

### 🎯 Interview Question
**Q: What is the difference between `find()` and `index()`?**
**Answer:** Both search for a substring. `find()` returns `-1` if not found. `index()` throws a `ValueError` and crashes your program if not found. In Data Science, `find()` is generally safer for messy data.

## 3. String Immutability
* **Definition:** Strings cannot be changed in place. Any modification creates a brand new string in memory.
* **Why?** Python caches and interns strings to save memory. If strings were mutable, changing one variable could accidentally change another variable pointing to the same cached string.
