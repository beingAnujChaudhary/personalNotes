# 📘 Module 1: Creating DataFrames

A **DataFrame** is a 2-dimensional labeled data structure with columns of potentially different types. It is the most widely used object in Pandas.

## 💡 Mental Model
If a **Series** is a single column in an Excel spreadsheet, a **DataFrame** is the **entire spreadsheet**. You can create it from lists of lists (row by row), dictionaries (column by column), or external files (CSV).

## 📊 Visual Diagram
```text
1. List of Lists (Row-wise)          2. Dictionary (Column-wise)
   [ [1, 2], [3, 4] ]                  {'A': [1, 3], 'B': [2, 4]}
         ↓                                     ↓
   ┌───┬───┐                             ┌───┬───┐
   │ 0 │ 1 │                             │ A │ B │
   ├───┼───┤                             ├───┼───┤
   │ 1 │ 2 │                             │ 1 │ 2 │
   │ 3 │ 4 │                             │ 3 │ 4 │
   └───┴───┘                             └───┴───┘
```

## 🌍 Real-World Use Case
Reading a 50,000-row CSV file of customer transactions into a DataFrame so you can analyze spending patterns, filter by date, and calculate total revenue.

## ⚠️ Common Mistakes
When passing a dictionary, all lists (columns) **must be of the same length**. If one is shorter, Pandas will throw a `ValueError`.

## 🎤 Interview Questions
**Q: What happens if you pass a dictionary to `pd.DataFrame()` where one list is shorter than the others?**
**A:** Pandas will raise a `ValueError: All arrays must be of the same length`. You must ensure all columns have the same number of rows.

# 📘 Module 2: Attributes & Exploratory Methods

**Attributes** (no parentheses) give you metadata about the DataFrame. **Methods** (with parentheses) perform actions like viewing data, summarizing statistics, or checking for missing values.

## 💡 Mental Model
Attributes are the DataFrame's **ID Card** (How big is it? What types are the columns?). Methods are the DataFrame's **Tools** (Show me the top 5 rows! Summarize the math!).

## 📊 Visual Diagram
```text
DataFrame
 ├── Attributes (No parentheses)
 │    ├── shape: (rows, cols)
 │    ├── dtypes: data types per column
 │    └── columns: list of column names
 └── Methods (Parentheses)
      ├── info(): memory usage & non-null counts
      ├── describe(): statistical summary (mean, std, min, max)
      └── isnull().sum(): count missing values per column
```

## 🌍 Real-World Use Case
Before training a Machine Learning model, you run `df.info()` and `df.isnull().sum()` to ensure you have enough data and to identify which columns need missing-value imputation.

## ⚠️ Common Mistakes
Forgetting that `describe()` only works on **numeric** columns by default. To see summaries for strings/objects, you must pass `include='all'`.

## 🎤 Interview Questions
**Q: What is the difference between `df.shape` and `df.info()`?**
**A:** `df.shape` returns a simple tuple `(rows, columns)`. `df.info()` prints a detailed summary including column names, non-null counts, data types, and total memory usage.

## 🎯 Practice Tasks

🟡 **Medium:** Check how many missing values exist in every column of `ipl`.

🔴 **Hard:** Rename the `iq` column in `students` to `IQ_Score` and the `package` column to `LPA`.


````python
# ==========================================
# 💻 CODE EXAMPLE & EXPLANATION
# ==========================================
print(f"Shape: {movies.shape}")
print(f"Columns: {movies.columns.tolist()}")
print(f"Dtypes:\n{movies.dtypes}\n")

# info() and describe()
print("--- Info ---")
ipl.info()

print("\n--- Describe ---")
print(movies.describe())

# Checking missing values and duplicates
print(f"\nNulls per column:\n{movies.isnull().sum()}")
print(f"\nDuplicate rows: {movies.duplicated().sum()}")

# Renaming columns
students.rename(columns={'iq': 'IQ_Score', 'package': 'LPA'}, inplace=True)
print(f"\nRenamed Columns: {students.columns.tolist()}")
````

# 📘 Module 3: Mathematical Methods & The `axis` Parameter

Pandas allows you to perform aggregate math (`sum`, `mean`, `var`) across the entire DataFrame. The most crucial concept here is the **`axis`** parameter.

## 💡 Mental Model
- **`axis=0`** means "pointing DOWN". It collapses rows together, applying the math to **each column**.
- **`axis=1`** means "pointing SIDEWAYS". It collapses columns together, applying the math to **each row**.

## 📊 Visual Diagram
```text
       Col1  Col2
Row1 [ 10,   20 ]  -> axis=1 (sum=30)
Row2 [ 30,   40 ]  -> axis=1 (sum=70)
       ↓      ↓
    axis=0   axis=0
   (sum=40) (sum=60)
```

## 🌍 Real-World Use Case
Calculating the **total marks per student** (sum across columns, `axis=1`) vs calculating the **class average per subject** (mean across rows, `axis=0`).

## ⚠️ Common Mistakes
Confusing `axis=0` and `axis=1`. Remember: `axis=0` operates *down* the index (rows), `axis=1` operates *across* the columns.

## 🎤 Interview Questions
**Q: If you run `df.sum(axis=1)`, what does the output represent?**
**A:** It returns a Series where each value is the sum of the columns for that specific row. The length of the output Series equals the number of rows in the DataFrame.

## 🎯 Practice Tasks

🟢 **Easy:** Calculate the sum of all numeric columns in `students` (default axis).

🟡 **Medium:** Calculate the mean of `iq` and `marks` for **each student** (row-wise).

🔴 **Hard:** Calculate the variance (`var()`) of the numeric columns in `students` across columns.


````python
# ==========================================
# 💻 CODE EXAMPLE & EXPLANATION
# ==========================================
# Reset students columns for this example
students.rename(columns={'IQ_Score': 'iq', 'LPA': 'package'}, inplace=True)

# axis=0 (Default) -> Sum of each column
print("--- Sum of Columns (axis=0) ---")
print(students.sum(axis=0))

# axis=1 -> Sum of each row
print("\n--- Sum of Rows (axis=1) ---")
print(students.sum(axis=1))

# Mean across rows
print("\n--- Mean of IQ and Marks per student ---")
print(students[['iq', 'marks']].mean(axis=1))
````

# 📘 Module 4: Selecting Rows & Columns (`loc` vs `iloc`)

You can extract specific columns using brackets `[]`. To extract rows, or both rows and columns simultaneously, you use **`iloc`** (integer position) and **`loc`** (label-based).

## 💡 Mental Model
- **`iloc`** is like a Python List. It uses **0-based integer positions**. (e.g., "Give me the 3rd row").
- **`loc`** is like a Python Dictionary. It uses the **actual index labels**. (e.g., "Give me the row named 'nitish'").

## 📊 Visual Diagram
```text
       Col1  Col2
RowA [ 10,   20 ]
RowB [ 30,   40 ]

iloc[0, 1] -> 20 (Position 0, Position 1)
loc['RowA', 'Col2'] -> 20 (Label 'RowA', Label 'Col2')
```

## 🌍 Real-World Use Case
Extracting the `Team1` and `Team2` columns for the first 5 matches (`iloc[:5, [2,3]]`) to display a quick preview on a dashboard.

## ⚠️ Common Mistakes
Using `loc` with integer positions when your index is strings. If your index is `['a', 'b', 'c']`, `loc[0]` will throw a `TypeError` because `0` is not a label!

## 🎤 Interview Questions
**Q: What is the difference in slicing behavior between `loc` and `iloc`?**
**A:** `iloc` slicing is **exclusive** of the end index (like standard Python lists: `0:3` gets 0, 1, 2). `loc` slicing is **inclusive** of the end label (`'a':'c'` gets 'a', 'b', and 'c').

## 🎯 Practice Tasks

🟢 **Easy:** Extract the `title_x` and `imdb_rating` columns from `movies`.

🟡 **Medium:** Extract the 2nd and 4th rows of `movies` using `iloc`.

🔴 **Hard:** Extract rows 'nitish' to 'rupesh' (inclusive) and only the `iq` and `marks` columns using `loc`.


````python
# ==========================================
# 💻 CODE EXAMPLE & EXPLANATION
# ==========================================
# 1. Selecting Columns
print("--- Single Column ---")
print(movies['title_x'].head(2))

print("\n--- Multiple Columns ---")
print(movies[['title_x', 'imdb_rating']].head(2))

# 2. Selecting Rows with iloc (Positions)
print("\n--- iloc (Positions) ---")
print("Single row:\n", movies.iloc[0])
print("Fancy indexing:\n", movies.iloc[[0, 2, 4]])

# 3. Selecting Rows with loc (Labels)
print("\n--- loc (Labels) ---")
print("Single row:\n", students.loc['nitish'])
print("Slicing (Inclusive!):\n", students.loc['nitish':'rupesh'])

# 4. Selecting Both Rows and Cols
print("\n--- Both Rows & Cols ---")
print("iloc[0:2, 0:2]:\n", movies.iloc[0:2, 0:2])
print("loc['nitish':'ankit', 'iq':'marks']:\n", students.loc['nitish':'ankit', 'iq':'marks'])
````

# 📘 Module 5: Filtering DataFrames (Boolean Indexing)

Filtering allows you to extract rows that meet specific conditions. You create a **Boolean Mask** (a Series of True/False) and pass it inside the DataFrame brackets `df[mask]`.

## 💡 Mental Model
Think of the mask as a **bouncer at a club**. The condition checks every row. If `True`, the row gets in. If `False`, it's filtered out. For multiple conditions, use `&` (AND), `|` (OR), and `~` (NOT).

## 📊 Visual Diagram
```text
Data:  [10, 50, 100]
Mask:  [F,  T,  T ]  (Data > 40)
Result:[50, 100]

Multiple Conditions:
(Data > 40) & (Data < 80) -> [F, T, F] & [T, F, F] -> [F, F, F]
```

## 🌍 Real-World Use Case
Finding all IPL finals where CSK won in Kolkata: `ipl[(ipl['WinningTeam'] == 'CSK') & (ipl['City'] == 'Kolkata')]`.

## ⚠️ Common Mistakes
Using Python's `and` / `or` instead of Pandas' `&` / `|`. Also, forgetting to wrap conditions in parentheses: `(df['A'] > 5) & (df['B'] < 10)`.

## 🎤 Interview Questions
**Q: How do you filter a DataFrame based on a partial string match in a column?**
**A:** Use the `.str.contains()` method. Example: `df[df['genres'].str.contains('Action')]`.

## 🎯 Practice Tasks

🟢 **Easy:** Filter `ipl` to show only the 'Final' matches.

🟡 **Medium:** Count how many Super Over finishes have occurred in `ipl`.

🔴 **Hard:** Filter `movies` to find Action movies with an `imdb_rating` > 7.5 using `.str.contains()`.


````python
# ==========================================
# 💻 CODE EXAMPLE & EXPLANATION
# ==========================================
# 1. Single Condition
finals = ipl[ipl['MatchNumber'] == 'Final']
print("--- IPL Finals ---")
print(finals[['Season', 'WinningTeam']])

# 2. Counting filtered rows
super_overs = ipl[ipl['SuperOver'] == 'Y'].shape[0]
print(f"\nSuper Over count: {super_overs}")

# 3. Multiple Conditions (&)
csk_kolkata = ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'CSK')]
print(f"\nCSK wins in Kolkata: {csk_kolkata.shape[0]}")

# 4. String Methods in Filtering
action_high_rated = movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating'] > 7.5)]
print("\n--- Action Movies > 7.5 Rating ---")
print(action_high_rated[['title_x', 'imdb_rating']])
````

# 📘 Module 6: Adding & Modifying Columns

DataFrames are mutable. You can add a completely new column with a constant value, or derive a new column by performing operations on existing columns.

## 💡 Mental Model
Adding a column is exactly like adding a new key-value pair to a dictionary. `df['new_col'] = ...`. If the right side is a single value, it broadcasts to all rows. If it's a Series, it aligns by index.

## 📊 Visual Diagram
```text
┌─────────┬──────┐      ┌─────────┬──────┬─────────┐
│ actors  │ ...  │  =>  │ actors  │ ...  │ lead    │
├─────────┼──────┤      ├─────────┼──────┼─────────┤
│ A|B     │ ...  │      │ A|B     │ ...  │ A       │
│ C|D     │ ...  │      │ C|D     │ ...  │ C       │
└─────────┴──────┘      └─────────┴──────┴─────────┘
                          (Split string, take 1st element)
```

## 🌍 Real-World Use Case
Extracting the "Lead Actor" from a column formatted as `"Actor 1|Actor 2"` by splitting the string and taking the first element.

## ⚠️ Common Mistakes
Using `df.column_name = values` to add a new column. This creates an *attribute*, not a column! Always use bracket notation: `df['column_name'] = values`.

## 🎤 Interview Questions
**Q: How do you extract the first element of a pipe-separated string in a Pandas column?**
**A:** Use `.str.split('|').apply(lambda x: x[0])` or `.str.split('|').str[0]`.

## 🎯 Practice Tasks

🟢 **Easy:** Add a new column `Country` to `movies` with the constant value `'India'`.

🟡 **Medium:** Create a new column `total_runs` in a mock DataFrame by adding two existing columns.

🔴 **Hard:** Extract the lead actor from the `actors` column in `movies` into a new column `lead_actor`.


````python
# ==========================================
# 💻 CODE EXAMPLE & EXPLANATION
# ==========================================
# 1. Adding a constant column
movies['Country'] = 'India'
print("--- Added Constant Column ---")
print(movies[['title_x', 'Country']].head(2))

# 2. Deriving from existing columns (String manipulation)
# Split by '|', then apply a lambda to get the first element (index 0)
movies['lead actor'] = movies['actors'].str.split('|').apply(lambda x: x[0])
print("\n--- Extracted Lead Actor ---")
print(movies[['title_x', 'actors', 'lead actor']].head(3))
````

# 📘 Module 7: Important Functions (`astype`, `value_counts`, `sort_values`)

- **`astype()`**: Converts a column to a different data type.
- **`value_counts()`**: Returns the frequency of unique values in a Series.
- **`sort_values()`**: Sorts the DataFrame by one or more columns.

## 💡 Mental Model
- `astype` changes the **lens** through which Pandas sees the data (e.g., treating numbers as text).
- `value_counts` is a **tally counter** for categorical data.
- `sort_values` is a **deck of cards** being organized by rank or suit.

## 📊 Visual Diagram
```text
['1', '2', '3'] -> astype(int) -> [1, 2, 3]

['A', 'B', 'A'] -> value_counts() -> A: 2, B: 1

[3, 1, 2] -> sort_values() -> [1, 2, 3]
```

## 🌍 Real-World Use Case
Using `value_counts()` to find which player won the most "Player of the Match" awards. Using `sort_values()` to rank movies by `imdb_rating` descending, and then by `year_of_release` ascending.

## ⚠️ Common Mistakes
Forgetting that `astype()` returns a *new* Series. You must assign it back: `df['col'] = df['col'].astype(int)`. Also, forgetting `ascending=False` when you want the highest values at the top.

## 🎤 Interview Questions
**Q: How do you sort a DataFrame by multiple columns with different ascending/descending orders?**
**A:** Pass lists to the parameters: `df.sort_values(by=['col1', 'col2'], ascending=[True, False])`.

## 🎯 Practice Tasks

🟢 **Easy:** Find the frequency of each `City` in the `ipl` DataFrame using `value_counts()`.

🟡 **Medium:** Sort the `movies` DataFrame by `imdb_rating` in descending order.

🔴 **Hard:** Convert the `year_of_release` column in `movies` to a string type using `astype()`.


````python
# ==========================================
# 💻 CODE EXAMPLE & EXPLANATION
# ==========================================
# 1. value_counts
print("--- City Frequencies ---")
print(ipl['City'].value_counts())

# 2. sort_values
print("\n--- Movies Sorted by Rating (Desc) ---")
print(movies.sort_values(by='imdb_rating', ascending=False)[['title_x', 'imdb_rating']])

# 3. astype
print("\n--- Data Type Conversion ---")
print(f"Before: {movies['year_of_release'].dtype}")
movies['year_of_release'] = movies['year_of_release'].astype(str)
print(f"After: {movies['year_of_release'].dtype}")
````
