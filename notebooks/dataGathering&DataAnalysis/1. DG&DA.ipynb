### 🔄 The ETL Process

ETL is the backbone of data engineering and is **extremely important for interviews**.

*   **Extract:** Pull data from source systems (APIs, DBs, Files).
*   **Transform:** Clean, aggregate, and format data (Remove duplicates, fix types).
*   **Load:** Insert into a target destination (Data Warehouse, SQL DB).

**Simple ETL Visualization:**
```text
CSV File
   │
Extract (Read CSV)
   │
Transform (Remove Duplicates, Fix Types)
   │
Load (Into SQL Database)
```

> 💡 **Fun Fact:** The term "Data Warehouse" was coined by Bill Inmon in the 1980s, but the concept of centralized data storage dates back to the 1960s with early mainframe systems!

````python

# ==========================================
# 🔄 ETL PIPELINE EXAMPLE
# ==========================================

# 1. EXTRACT (Simulating reading from a messy CSV)
csv_data = """order_id,customer_id,amount,date
1,101,250.00,2023-10-01
2,102,150.50,2023-10-01
1,101,250.00,2023-10-01  # Duplicate!
3,103,,2023-10-02        # Missing amount!
"""

# FIX: Added comment='#' so Pandas ignores the inline notes
df_extract = pd.read_csv(StringIO(csv_data), comment='#')

print("--- EXTRACT: Raw Data ---")
print(df_extract)

# 2. TRANSFORM
print("\n--- TRANSFORM: Cleaning Data ---")

# Check for duplicates
print(f"Duplicates found: {df_extract.duplicated().sum()}")
df_transform = df_extract.drop_duplicates()

# Check for missing values
print(f"Missing values:\n{df_transform.isnull().sum()}")
df_transform['amount'] = df_transform['amount'].fillna(0)

# Convert data types
# FIX: Use .str.strip() to remove the hidden trailing spaces before converting
df_transform['date'] = pd.to_datetime(df_transform['date'].str.strip())

print("\nCleaned Data:")
print(df_transform)
# 3. LOAD (Into an in-memory SQLite Database)
conn = sqlite3.connect(':memory:')
df_transform.to_sql('cleaned_orders', conn, if_exists='replace', index=False)

# Verify Load
df_loaded = pd.read_sql('SELECT * FROM cleaned_orders', conn)
print("\n--- LOAD: Verified Data from SQL ---")
display(df_loaded)
````

### 🧠 Memory Optimization for Large Files (Interview Favorite)
**Problem**: You have a 10 GB CSV file but only 8 GB of RAM. `pd.read_csv()` will crash.
- **Solution 1 (Chunk Processing)**: Read the file in chunks (e.g., 100,000 rows at a time).
- **Solution 2 (Memory Optimization)**: Pandas uses 64-bit floats and ints by default, which wastes RAM. You can reduce memory usage by **up to 90%** by specifying `dtype`.

### 📦 Chunk Processing
If a file is larger than your RAM, use `chunksize`. This returns an iterator, allowing you to process the file in manageable blocks.

**Problem:** 10 GB CSV File, 8 GB RAM → Crash ❌

**Solution:** Chunk Processing ✓


````python
# ==========================================
# 📥 MEMORY OPTIMIZATION & CHUNKING
# ==========================================

# 1. Memory Optimization with dtypes
dtypes = {
    'id': 'int32',         # Instead of default int64
    'category': 'category' # Pandas categorical saves massive memory
}
# df_optimized = pd.read_csv('huge_file.csv', dtype=dtypes)

print("📊 Memory Optimization Example:")
df_demo = pd.DataFrame({'id': range(1000), 'category': ['A']*500 + ['B']*500})
print(f"Default memory: {df_demo.memory_usage(deep=True).sum() / 1024:.2f} KB")

df_demo['id'] = df_demo['id'].astype('int32')
df_demo['category'] = df_demo['category'].astype('category')
print(f"Optimized memory: {df_demo.memory_usage(deep=True).sum() / 1024:.2f} KB")

# 2. Chunk Processing (For files larger than RAM)
print("\n📦 Chunk Processing Example:")
chunk_size = 5

# Simulate reading in chunks
sample_data = pd.DataFrame({'id': range(15), 'value': np.random.rand(15)})
for i in range(0, len(sample_data), chunk_size):
    chunk = sample_data.iloc[i:i+chunk_size]
    print(f"Processing chunk {i//chunk_size + 1}: shape {chunk.shape}")

# Real chunking code:
# chunk_iter = pd.read_csv('large_file.csv', chunksize=10000)
# for chunk in chunk_iter:
#     processed = chunk[chunk['value'] > 100]
#     processed.to_csv('output.csv', mode='a', header=False)
````

### 4.3 Import Data from JSON

JSON is a lightweight data-interchange format that's easy for humans to read and write. It supports nested and hierarchical data structures.


````python
# 📊 Import Data from JSON

import json

# Basic JSON import
df = pd.read_json('filename.json')

# Read JSON with specific orientation
df = pd.read_json('filename.json', orient='records')  # List of records
df = pd.read_json('filename.json', orient='columns')  # Dict of columns
df = pd.read_json('filename.json', orient='index')    # Dict of indices

# Read JSON Lines (one JSON object per line)
df = pd.read_json('filename.jsonl', lines=True)

# Handle nested JSON with json_normalize
from pandas import json_normalize

nested_data = {
    "company": "TechCorp",
    "employees": [
        {"name": "John", "skills": ["Python", "SQL"], "address": {"city": "NYC", "zip": "10001"}},
        {"name": "Jane", "skills": ["Java", "AWS"], "address": {"city": "SF", "zip": "94105"}}
    ]
}

# Flatten nested JSON
df_normalized = json_normalize(
    nested_data['employees'],
    sep='_'  # Separator for nested keys
)
print("Flattened Nested JSON:")
print(df_normalized)
````

#### 🕸️ Nested JSON Problem

APIs return JSON, which is often deeply nested. A standard DataFrame import will leave dictionaries inside your columns. Use `json_normalize()` to flatten the hierarchy.

````python
# 🎯 Advanced JSON Manipulation Example

# Sample nested JSON data
json_data = '''
{
    "store": "Electronics Plus",
    "location": "Downtown",
    "products": [
        {
            "id": 1,
            "name": "Laptop",
            "price": 999.99,
            "specs": {"ram": "16GB", "storage": "512GB SSD"},
            "reviews": [
                {"user": "Alice", "rating": 5, "comment": "Great!"},
                {"user": "Bob", "rating": 4, "comment": "Good value"}
            ]
        },
        {
            "id": 2,
            "name": "Mouse",
            "price": 29.99,
            "specs": {"type": "Wireless", "dpi": 1600},
            "reviews": [
                {"user": "Charlie", "rating": 3, "comment": "Average"}
            ]
        }
    ]
}
'''

# Parse JSON
data = json.loads(json_data)

# Normalize main data with nested specs
df_products = json_normalize(data['products'], sep='_')
print("📦 Product Data:")
print(df_products)

# Extract reviews (nested arrays)
reviews_list = []
for product in data['products']:
    for review in product['reviews']:
        reviews_list.append({
            'product_id': product['id'],
            'product_name': product['name'],
            'user': review['user'],
            'rating': review['rating'],
            'comment': review['comment']
        })

df_reviews = pd.DataFrame(reviews_list)
print("\n⭐ Reviews Data:")
print(df_reviews)
````

### 4.4 Import Data from SQL Databases

Most company data exists in SQL databases. Python provides multiple ways to connect and query them.
#### 🗄️ Why SQL is Important

Most company data exists in databases (PostgreSQL, SQL Server, Snowflake, BigQuery). Knowing how to use `pd.read_sql()` with complex `JOIN` queries is essential.

````python
# 📊 Import Data from SQL Databases

import sqlite3
from sqlalchemy import create_engine

# Method 1: Using SQLite (built-in Python)
conn = sqlite3.connect('database.db')
df = pd.read_sql_query("SELECT * FROM table_name", conn)

# Method 2: Using SQLAlchemy (for production databases)
# MySQL
engine_mysql = create_engine('mysql+pymysql://username:password@localhost:3306/dbname')
# PostgreSQL
engine_pg = create_engine('postgresql://username:password@localhost:5432/dbname')
# SQLite
engine_sqlite = create_engine('sqlite:///database.db')

df = pd.read_sql_query("SELECT * FROM table_name", engine_pg)

# Read with complex JOIN query
query = """
    SELECT 
        e.name,
        d.department_name,
        e.salary,
        e.hire_date
    FROM employees e
    JOIN departments d ON e.dept_id = d.id
    WHERE e.salary > 70000
    ORDER BY e.salary DESC
"""
df = pd.read_sql_query(query, conn)

# Read with parameters (prevents SQL injection)
query = "SELECT * FROM employees WHERE department = :dept AND salary > :sal"
df = pd.read_sql_query(query, conn, params={'dept': 'IT', 'sal': 70000})

# Chunked reading for large tables
chunks = []
for chunk in pd.read_sql("SELECT * FROM large_table", conn, chunksize=5000):
    processed = chunk[chunk['value'] > 100]
    chunks.append(processed)
df_large = pd.concat(chunks, ignore_index=True)

conn.close()
````

````python
# 🎯 Complete SQL Database Example

# Create SQLite database and tables
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

# Create departments table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS departments (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL
    )
''')

# Create employees table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS employees (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER,
        department_id INTEGER,
        salary REAL,
        FOREIGN KEY (department_id) REFERENCES departments(id)
    )
''')

# Insert sample data
cursor.execute("INSERT INTO departments VALUES (1, 'IT'), (2, 'HR'), (3, 'Finance')")
cursor.execute("INSERT INTO employees VALUES (1, 'Alice', 25, 1, 70000)")
cursor.execute("INSERT INTO employees VALUES (2, 'Bob', 30, 2, 65000)")
cursor.execute("INSERT INTO employees VALUES (3, 'Charlie', 35, 1, 80000)")
cursor.execute("INSERT INTO employees VALUES (4, 'Diana', 28, 3, 75000)")
conn.commit()

# Query 1: Simple SELECT
df_employees = pd.read_sql_query("SELECT * FROM employees", conn)
print("👥 All Employees:")
print(df_employees)

# Query 2: Aggregation with GROUP BY
query_agg = """
    SELECT 
        department_id,
        COUNT(*) as emp_count,
        AVG(salary) as avg_salary,
        MAX(salary) as max_salary,
        MIN(salary) as min_salary
    FROM employees
    GROUP BY department_id
"""
df_agg = pd.read_sql_query(query_agg, conn)
print("\n📊 Department Statistics:")
print(df_agg)

# Query 3: JOIN example
query_join = """
    SELECT 
        e.name as employee_name,
        e.age,
        d.name as department,
        e.salary
    FROM employees e
    JOIN departments d ON e.department_id = d.id
    WHERE e.salary > 70000
    ORDER BY e.salary DESC
"""
df_joined = pd.read_sql_query(query_join, conn)
print("\n🔗 High-Earning Employees with Departments:")
print(df_joined)

conn.close()
````

## 📤 Section 5: Exporting Data & Format Comparison

### 📊 Data Format Comparison Table

| Format | Best Use Case | Pros | Cons |
| :--- | :--- | :--- | :--- |
| **CSV** | Small/Medium Data | Universal, human-readable | No types, no formulas, slow for big data |
| **Excel** | Business Reports | Multiple sheets, formatting | Max 1M rows, slow, binary |
| **JSON** | APIs, NoSQL, Web Apps | Nested data, web-native | Verbose, large file size, slow to parse |
| **SQL** | Large Structured Data | Relational, fast queries, secure | Requires DB setup infrastructure |
| **HTML** | Email/Web Reports | Browser-viewable | Not for data analysis, hard to parse programmatically |
| **Parquet** | Big Data / ML | Columnar, highly compressed, fast | Not human-readable without code|

> 💡 **Pro Tip:** Always use `index=False` when exporting to CSV/Excel unless your index contains meaningful business keys (like `customer_id`). Otherwise, you'll export an ugly `Unnamed: 0` column!

## 🌐 Section 6: Advanced API Integration

### 🌐 What is an API?

**API (Application Programming Interface)** allows applications to communicate with each other and exchange data.

### 🔄 API Architecture

```text
Client (Python) ──(HTTP GET Request + Auth)──▶ API Gateway ──▶ Database
Client (Python) ◀──(JSON Response)───────── API Gateway
```

*   **Authentication:** Usually passed via Headers (`Authorization: Bearer <token>`).
*   **Rate Limiting:** APIs restrict requests (e.g., 100/min). Exceeding this returns `429 Too Many Requests`.
*   **Solution:** Implement **Exponential Backoff** or respect the `Retry-After` header.

````python
# 🌐 ADVANCED API: AUTH & RATE LIMITS

def fetch_api_data(url, headers=None, max_retries=3):
    """Robust API fetcher with rate limit handling and exponential backoff"""
    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=10)
            
            # Handle Rate Limiting (429 Too Many Requests)
            if response.status_code == 429:
                retry_after = int(response.headers.get('Retry-After', 60))
                print(f"⚠️ Rate limited. Sleeping for {retry_after}s...")
                time.sleep(retry_after)
                continue
                
            response.raise_for_status()
            return response.json()
            
        except requests.exceptions.RequestException as e:
            print(f"❌ Attempt {attempt+1} failed: {e}")
            time.sleep(2 ** attempt)  # Exponential backoff: 1s, 2s, 4s
            
    return None

# Example Usage
# headers = {"Authorization": "Bearer YOUR_TOKEN"}
# data = fetch_api_data("https://api.example.com/data", headers=headers)
print("✅ Robust API Fetcher defined with rate limit handling!")

# Different Authentication Methods
print("\n📋 Authentication Methods:")
print("1. API Key in Header: headers = {'Authorization': 'Bearer YOUR_API_KEY'}")
print("2. API Key in Query: params = {'api_key': 'YOUR_API_KEY'}")
print("3. Basic Auth: auth=HTTPBasicAuth('username', 'password')")
print("4. OAuth 2.0: Uses access tokens with refresh mechanism")
````

````python
# 📊 API Data Processing Pipeline Class

class APIDataCollector:
    """A class to handle API data collection and processing"""
    
    def __init__(self, base_url, api_key=None):
        self.base_url = base_url
        self.api_key = api_key
        self.session = requests.Session()
        if api_key:
            self.session.headers.update({'Authorization': f'Bearer {api_key}'})
    
    def fetch_data(self, endpoint, params=None):
        """Fetch data from API endpoint with error handling"""
        url = f"{self.base_url}/{endpoint}"
        
        try:
            response = self.session.get(url, params=params)
            response.raise_for_status()
            
            # Handle rate limiting
            if response.status_code == 429:
                retry_after = int(response.headers.get('Retry-After', 60))
                print(f"Rate limited. Waiting {retry_after}s...")
                time.sleep(retry_after)
                return self.fetch_data(endpoint, params)
                
            return response.json()
        except requests.exceptions.RequestException as e:
            print(f"❌ API Error: {e}")
            return None
    
    def process_to_dataframe(self, data, columns=None):
        """Convert API response to DataFrame"""
        if data is None:
            return None
        
        if isinstance(data, list):
            df = pd.DataFrame(data)
        elif isinstance(data, dict):
            if 'results' in data:
                df = pd.DataFrame(data['results'])
            else:
                df = pd.DataFrame([data])
        else:
            return None
        
        if columns:
            df = df[[col for col in columns if col in df.columns]]
        
        return df
    
    def save_data(self, df, filename, format='csv'):
        """Save DataFrame to file"""
        if format == 'csv':
            df.to_csv(f"{filename}.csv", index=False)
        elif format == 'json':
            df.to_json(f"{filename}.json", orient='records')
        elif format == 'excel':
            df.to_excel(f"{filename}.xlsx", index=False)
        print(f"✅ Data saved to {filename}.{format}")
    
    def close(self):
        """Close the session"""
        self.session.close()

print("✅ APIDataCollector class ready for production use!")
````

## 🕸️ Section 7: Advanced Web Scraping

### 🕸️ What is Web Scraping?

**Web Scraping** is the process of extracting data from websites automatically using software.

### ⚖️ Legal and Ethical Considerations

Always check before scraping:
1. **Legal:** Always check `website.com/robots.txt` and Terms of Service. Never scrape PII.
2. **The Problem:** Modern sites use JavaScript to load data dynamically. `requests.get()` will return empty HTML.
3. **The Solution:** Use Selenium or Playwright to automate a real browser that executes JavaScript.

### 🔄 Scraping Workflow

```text
Website HTML
   │
Fetch Page (requests.get())
   │
Parse HTML (BeautifulSoup)
   │
Extract Data (Tags: find(), select())
   │
Pandas DataFrame
```

### 🚫 The Dynamic Website Problem

Modern sites use React/Angular. The HTML returned by `requests` is empty because JavaScript populates the data *after* load.

**Solution:** Use browser automation tools like **Selenium** or **Playwright**.


````python
# 🕸️ Web Scraping - Basic Examples

from bs4 import BeautifulSoup
import requests

# Basic web scraping
url = 'https://quotes.toscrape.com/'
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract quotes
    quotes = []
    for quote_div in soup.find_all('div', class_='quote'):
        text = quote_div.find('span', class_='text').text
        author = quote_div.find('small', class_='author').text
        tags = [tag.text for tag in quote_div.find_all('a', class_='tag')]
        
        quotes.append({
            'quote': text,
            'author': author,
            'tags': ', '.join(tags)
        })
    
    df_quotes = pd.DataFrame(quotes)
    print(f"✅ Scraped {len(df_quotes)} quotes!")
    print(df_quotes.head())
else:
    print(f"❌ Failed to access website: {response.status_code}")
````

````python
# 🛠️ Advanced Web Scraper Class

class EthicalScraper:
    """Advanced web scraping class with ethical practices"""
    
    def __init__(self, base_url, delay=1):
        self.base_url = base_url
        self.delay = delay
        self.session = requests.Session()
        self.session.headers.update({
            'User-Agent': 'Mozilla/5.0 (compatible; DataCollection/1.0; +http://example.com/bot)'
        })
    
    def fetch_page(self, url):
        """Fetch a page with error handling and rate limiting"""
        try:
            time.sleep(self.delay)
            response = self.session.get(url, timeout=10)
            response.raise_for_status()
            return BeautifulSoup(response.content, 'html.parser')
        except Exception as e:
            print(f"❌ Error fetching {url}: {e}")
            return None
    
    def extract_table(self, soup, table_id=None, table_class=None):
        """Extract HTML table to DataFrame"""
        if table_id:
            table = soup.find('table', id=table_id)
        elif table_class:
            table = soup.find('table', class_=table_class)
        else:
            table = soup.find('table')
        
        if not table:
            return None
        
        # Extract headers
        headers = [th.text.strip() for th in table.find_all('th')]
        
        # Extract rows
        rows = []
        for tr in table.find_all('tr')[1:]:  # Skip header row
            cells = [td.text.strip() for td in tr.find_all('td')]
            if cells:
                rows.append(cells)
        
        return pd.DataFrame(rows, columns=headers if headers else None)
    
    def paginate(self, url_pattern, start_page=1, end_page=10):
        """Handle pagination automatically"""
        all_data = []
        
        for page in range(start_page, end_page + 1):
            url = url_pattern.format(page)
            print(f"📄 Scraping page {page}...")
            
            soup = self.fetch_page(url)
            if soup is None:
                break
            
            # Extract data (customize based on site structure)
            items = soup.find_all('div', class_='item')
            for item in items:
                all_data.append({'text': item.text.strip()})
        
        return pd.DataFrame(all_data)
    
    def close(self):
        """Close session"""
        self.session.close()

print("✅ EthicalScraper class ready for production use!")
````

- `read_sql()`: Reads data FROM a SQL database INTO a DataFrame
- `to_sql()`: Writes a DataFrame TO a SQL database table

2. **Respect `Retry-After` header:** Use the wait time the API provides
3. **Request queuing:** Limit concurrent requests to stay under limits
4. **Response caching:** Cache frequent requests to avoid redundant calls
5. **Rotating API keys/Proxies:** Distribute load across multiple credentials

**Q16: What are Data Lakes and Data Warehouses?**
- **Data Lake:** Stores raw, unprocessed data in native format (S3, Hadoop). Schema-on-read.
- **Data Warehouse:** Stores processed, structured data for analytics (Snowflake, BigQuery). Schema-on-write.
- **Key difference:** Lakes are flexible and cheap; Warehouses are fast and structured.

**Q17: How do you scrape JavaScript-heavy websites?**
Three approaches:
1. **Use Selenium/Playwright:** Automate a real browser to execute JavaScript
2. **Find hidden APIs:** Open Chrome DevTools → Network tab → Find the XHR/Fetch calls the JS makes, then call those APIs directly with `requests`
3. **Use headless browsers:** Run Chrome/Firefox in headless mode for server-side scraping

**Q18: How would you collect data from 10 different sources and combine it?**
I would build a modular ETL pipeline:
1. Create individual extraction modules for each source type (CSV, Excel, API, DB, Scraping)
2. Standardize all data to a common schema (same column names, data types)
3. Implement data quality checks at each stage
4. Use a staging area to validate before final merge
5. Track lineage and version data for reproducibility
