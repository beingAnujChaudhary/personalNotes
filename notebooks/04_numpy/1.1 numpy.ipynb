# 🚀 Concept 1: What is NumPy & Basic Array Creation

**NumPy** (Numerical Python) is the foundation of scientific computing in Python. At its core is the `ndarray` (n-dimensional array), which stores **homogeneous** data in contiguous memory blocks.


### ⚡ Why NumPy over Python Lists?
| Feature | Python List | NumPy Array |
|--------|-------------|-------------|
| **Memory** | Non-contiguous, stores object pointers | Contiguous, stores raw data |
| **Data Types** | Heterogeneous (mixed) | Homogeneous (single type) |
| **Performance** | Slower (Python-level loops) | C-optimized, vectorized operations |
| **Ecosystem** | General purpose | Foundation for Pandas, Scikit-Learn, PyTorch |

### 📊 Memory Diagram: List vs NumPy Array
```text
Python List: Scattered Memory (Slow)
[ ptr ] -> [ int obj ]
[ ptr ] -> [ float obj ]
[ ptr ] -> [ str obj ]

NumPy Array: Contiguous Memory (Fast)
[ 1 | 2 | 3 | 4 | 5 ] -> Single C-array block
```

💡 **Fun Fact:** NumPy was created by Travis Oliphant in 2005. The heavy lifting is done in C and Fortran, making it up to **100x faster** than native Python loops!


````python
import numpy as np
import sys
import time

# ==========================================
# 💻 EXAMPLE: Speed & Memory Benchmark
# ==========================================

n = 10_000_000

# --- MEMORY COMPARISON ---
py_list = list(range(n))
# True memory of list = size of list object + size of all int objects
true_list_mem = sys.getsizeof(py_list) + sum(sys.getsizeof(i) for i in py_list)

np_arr = np.arange(n, dtype=np.int32)
np_mem = np_arr.nbytes

print(f"📋 Python List Memory: {true_list_mem / 1e6:.2f} MB")
print(f"🔢 NumPy int32 Memory: {np_mem / 1e6:.2f} MB")
print(f"💾 Memory Saved: {true_list_mem / np_mem:.1f}x smaller!\n")
````

````python
# --- SPEED COMPARISON ---
# Python List
start = time.time()
py_sq = [x**2 for x in py_list[:1000000]] # Using 1M for speed
list_time = time.time() - start

# NumPy
start = time.time()
np_sq = np_arr[:1000000] ** 2
np_time = time.time() - start

print(f"⏱️ Python List Time: {list_time:.4f} s")
print(f"⏱️ NumPy Time:       {np_time:.6f} s")
print(f"⚡ Speedup:          {list_time/np_time:.1f}x faster!")
````

# 🛠️ Concept 2: Specialized Array Creation

NumPy provides built-in functions to generate arrays without typing out every element.

### 📊 Visualizing `arange` vs `linspace`


````python
# np.arange(start, stop, step)
# Works like Python's built-in range(). 
# Starts at 0, steps by 2, and stops BEFORE reaching 10.
# Output: [0, 2, 4, 6, 8]

print("arange:", np.arange(0, 10, 2))
````

````python
# np.linspace(start, stop, num_points)
# Generates a specific NUMBER of evenly spaced points between two values.
# Starts at 0, ends exactly at 10 (inclusive by default), generating exactly 5 numbers.
# Output: [ 0.   2.5  5.   7.5 10. ]

print("linspace:", np.linspace(0, 10, 5))
````

### Pre-filled Arrays (Zeros and Ones)


````python
# np.zeros(shape)
# Creates a 2D matrix with 2 rows and 3 columns filled entirely with 0.0 (floats).
# Often used to initialize empty placeholders for data.

print("\nZeros:\n", np.zeros((2, 3)))
````

````python
# np.ones(shape)
# Creates a 2D matrix with 3 rows and 2 columns filled entirely with 1.0 (floats).

print("Ones:\n", np.ones((3, 2)))
````

````python
# np.random.randint(low, high, size)
# Generates random integers starting from 1 up to (but NOT including) 100.
# size=(2, 2) structures these random integers into a 2x2 grid.

print("Random Ints:", np.random.randint(1, 100, size=(2, 2)))
````

## 🎤 Interview Question
**Q: What is the key difference between `np.arange()` and `np.linspace()`?**
> **A:** `np.arange()` generates values based on a **step size** and excludes the stop value (like Python's `range`). `np.linspace()` generates a specific **number of points** and includes the stop value by default. `linspace` is preferred for plotting axes to avoid floating-point precision issues.

# 📊 Concept 3: Array Attributes & Type Casting

Every `ndarray` has metadata describing its structure and memory footprint.

| Attribute | Description | Example |
|-----------|-------------|---------|
| `ndim` | Number of dimensions (axes) | 1, 2, 3... |
| `shape` | Tuple indicating size of each dimension  | `(3, 4)` |
| `size` | Total number of elements | `12` |
| `itemsize` | Memory size of **one** element in bytes | `8` for `float64` |
| `dtype` | Data type of elements | `int32`, `float64` |

### Examining Array Metadata    

````python
# Create a 2D array (matrix) with 2 rows and 3 columns.
# We explicitly set the data type to 64-bit floats (float64).

arr = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.float64)
````

````python
# arr.ndim: Number of dimensions (axes) of the array.
# Since it's a matrix consisting of rows and columns, it is 2-dimensional.

print("ndim:", arr.ndim)
````

````python
# arr.shape: A tuple representing the exact size of each dimension.
# Output is (2, 3) representing 2 rows and 3 columns.

print("shape:", arr.shape)
````

````python
# arr.size: The total number of individual elements inside the array.
# 2 rows * 3 columns = 6 total elements.

print("size:", arr.size)
````

````python
# arr.itemsize: The memory size (in bytes) of a single element.
# Since we used float64, each number takes 64 bits. 
# Because 8 bits = 1 byte, 64 bits = 8 bytes per item.

print("itemsize:", arr.itemsize)
````

````python
# arr.dtype: The uniform data type of the array's elements.

print("dtype:", arr.dtype)
````

````python
# arr.nbytes: Total bytes consumed by the array in your system's RAM.
# Formula: size (6) * itemsize (8) = 48 bytes.

print("Total Memory:", arr.nbytes, "bytes")
````

#### 🔄 Concept 6: Type Casting (`astype`)
Sometimes, data is imported as highly precise `float64` numbers, but you only need standard integers. NumPy arrays have a fixed data type, meaning you cannot change them directly. Instead, you use `.astype()` to create a **new copy** of the array with the updated data type.

> 💡 **Pro Tip:** Downgrading from `float64` to `float32` or `int32` is a standard technique in Deep Learning to halve your RAM/VRAM usage with minimal loss of accuracy!

🔄 **Type Casting:** Use `arr.astype(new_dtype)` to change types. *Note: This always returns a **copy**, not a view!*


````python
# Type Casting and Memory Reduction 

# arr.astype(new_dtype) creates a NEW array where every element 
# is converted from the original float64 to a 32-bit integer (int32).
# Note: Converting floats to integers truncates any decimal values.

arr_int = arr.astype(np.int32)

print("Casted to int32:\n", arr_int, "\n")

# A 32-bit integer takes up exactly 32 bits of memory.
# 32 bits / 8 = 4 bytes per item.
# By casting, we successfully reduced the memory footprint per element by 50%!

print("New itemsize:", arr_int.itemsize, "bytes")
````

````python
sample = np.array([[1, 2, 3], [4, 5, 6]], dtype=np.float64)

print(f"Dimensions (ndim): {sample.ndim}")
print(f"Shape: {sample.shape}")
print(f"Total Elements (size): {sample.size}")
print(f"Data Type (dtype): {sample.dtype}")
print(f"Total Memory (bytes): {sample.nbytes}")

# Changing dtype to save memory
int_version = sample.astype(np.int16)
print(f"\nCasted to int16. New Memory: {int_version.nbytes} bytes")
````

# ⚡ Concept 4: Array Operations & Vectorization

NumPy supports **vectorized operations**, meaning you can perform mathematical operations (element-wise) on entire arrays without writing explicit Python `for` loops.

### 📊 Visualizing Operations

**Vector operations**: Element-wise math between arrays of the same shape.

```python
Scalar Op:  [1, 2, 3] + 10  =>  [11, 12, 13]
Vector Op:  
[1, 2, 3] * [2, 2, 2]  =>  [2, 4, 6]  (Element-wise!)
arr1 + arr2  # Element-wise addition
arr1 > arr2  # Element-wise comparison → returns a Boolean array
```

⚠️ **Crucial:** The `*` operator in NumPy is **element-wise multiplication**, NOT matrix multiplication!


````python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

# Scalar Operations
print("Add 5:", a + 5)
print("Square:", a ** 2)
print("Greater than 15:", a > 15) # Returns boolean array

# Vector Operations
print("\nElement-wise Add:", a + b)
print("Element-wise Multiply:", a * b)

# Matrix Multiplication (Dot Product)
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("\nElement-wise (*):\n", A * B)
print("Matrix Multiply (@):\n", A @ B) # or np.dot(A, B)
````

````python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Scalar Multiplication (* 2):", a * 2)
print("Vector Addition (a + b):", a + b)
print("Relational Comparison (a > 15):", a > 15)

# Masking: Filtering arrays based on conditions
mask = a > 15
print("Filtered Array:", a[mask])
````

# 📈 Concept 5: Array Functions & Aggregations

NumPy provides optimized math functions. Understanding the `axis` parameter is critical!

### 📊 The `axis` Paradigm
Understanding the `axis` parameter is critical for 2D arrays (matrices).
* `axis=0` → Operates vertically (Down columns, collapsing rows).
* `axis=1` → Operates horizontally (Across columns, collapsing columns per row).
```text
       axis=0 (Down / Columns)
         ↓   ↓   ↓
      [[ 1,  2,  3],  → axis=1 (Across / Rows)
       [ 4,  5,  6]]     → → →
```
**Rule of thumb:** 
1.`axis=0` collapses rows (operates vertically). 
2. `axis=1` collapses columns (operates horizontally).

**Key Functions:**
* `sum()`, `min()`, `max()`
* `mean()`, `median()`, `std()`

````python
mat = np.array([[1, 2, 3], 
                [4, 5, 6]])

# Aggregations
print("Total Sum:", mat.sum())
print("Col Max (axis=0):", mat.max(axis=0)) # [4, 5, 6]
print("Row Mean (axis=1):", mat.mean(axis=1)) # [2. 5.]

# Statistics
print("\nStd Dev:", mat.std())
print("Variance:", mat.var())

# Math Functions
angles = np.array([0, np.pi/2, np.pi])
print("\nSin:", np.sin(angles))

# Rounding
decimals = np.array([1.2, 2.6, -1.5])
print("\nFloor:", np.floor(decimals))
print("Ceil:", np.ceil(decimals))
print("Round:", np.round(decimals))
````

# 🎯 Concept 6: Indexing & Slicing

Extracting data using indices. 

**Crucial Rule:** Slicing returns a **VIEW**, not a copy! Modifying the slice modifies the original array.

* **Syntax:** `arr[start:stop:step]`
* **2D Syntax:** `arr[row_start:row_stop, col_start:col_stop]`
* ⚠️ **Crucial Note:** Slicing a NumPy array returns a **VIEW**, not a copy. Modifying a slice modifies the original array. Use `.copy()` if you want an isolated dataset.


````python
grid = np.arange(1, 26).reshape(5, 5)
print("Original Grid:\n", grid)

# Basic Indexing
print("\nElement at row 1, col 2:", grid[1, 2]) # 8

# Slicing
print("\nFirst 2 rows, last 2 cols:\n", grid[:2, 3:])

# Boolean Masking
print("\nEven numbers:\n", grid[grid % 2 == 0])

# VIEW vs COPY demonstration
slice_view = grid[0, :]
slice_view[0] = 999
print("\nGrid after modifying slice:\n", grid) # 999 is in the grid!
````

````python
grid = np.arange(1, 13).reshape(3, 4)
print("Original Grid:\n", grid)

print("\nElement at Row 1, Col 2:", grid[1, 2])
print("\nFirst 2 Rows, Last 2 Cols:\n", grid[:2, 2:])

# View vs Copy demonstration
slice_view = grid[0, :]
slice_view[0] = 999
print("\nOriginal array mutated by view:\n", grid)
````

# 🔄 Concept 7: Iteration, Reshaping & Flattening

While vectorization is preferred, sometimes you must iterate. For reshaping, remember that data is read in **C-contiguous (row-major)** order by default.

| Method | Description |
|--------|-------------|
| `.reshape(shape)` | Changes dimensions (e.g., 1D to 2D) |
| `.ravel()` | Flattens a multi-dimensional array to 1D |
| `np.vstack([a, b])` | Stacks arrays vertically (adds rows) |
| `np.hstack([a, b])` | Stacks arrays horizontally (adds columns) |

### 📊 Reshape & Transpose Visual
```text
Original (2x3)       Transpose (3x2)      Ravel (1x6)
[1, 2, 3]            [1, 4]               [1, 2, 3, 4, 5, 6]
[4, 5, 6]            [2, 5]
                     [3, 6]
```



````python
arr = np.arange(1, 7).reshape(2, 3)

# Iteration
print("Row by row:")
for row in arr:
    print(row)

print("\nElement by element (nditer):")
for item in np.nditer(arr):
    print(item, end=" ")

# Reshaping
print("\n\nReshape to 3x2:\n", arr.reshape(3, 2))
print("Transpose (.T):\n", arr.T)

# Ravel vs Flatten
ravel_view = arr.ravel()
flatten_copy = arr.flatten()
````

````python
# Reshaping
arr = np.arange(12)
grid = arr.reshape(3, 4)

# Stacking
a = np.array([[1, 2]])
b = np.array([[3, 4]])

print("Vertical Stack:\n", np.vstack([a, b]))
print("\nHorizontal Stack:\n", np.hstack([a, b]))
````

## 🎤 Interview Question
**Q: What is the difference between `ravel()` and `flatten()`?**
> **A:** `ravel()` returns a **view** of the original array whenever possible (no memory copy, highly efficient). `flatten()` always returns a **deep copy** (allocates new memory). Use `ravel()` for performance, `flatten()` when you need to modify the 1D array without affecting the original.


# 📦 Concept 8: Stacking & Splitting

Combine or divide arrays along specific axes.

### 📊 Visual Guide
```text
vstack (Vertical):      hstack (Horizontal):
[A]                     [A | B]
[B]

vsplit (Rows):          hsplit (Cols):
[Top]                   [Left | Right]
[Bottom]
```


````python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# Stacking
print("vstack:\n", np.vstack((a, b)))
print("\nhstack:\n", np.hstack((a, b)))

# Splitting
big = np.arange(1, 13).reshape(4, 3)
print("\nOriginal:\n", big)

left, right = np.hsplit(big, 2) # Split into 2 equal column parts
print("\nhsplit Left:\n", left)

top, bottom = np.vsplit(big, 2) # Split into 2 equal row parts
print("\nvsplit Top:\n", top)
````

## 🎤 Interview Question
**Q: What is the difference between `np.concatenate()`, `np.vstack()`, and `np.stack()`?**
> **A:** 
> - `np.concatenate()` joins a sequence of arrays along an **existing axis**.
> - `np.vstack()` and `np.hstack()` are convenience wrappers around `concatenate` specifically for vertical (axis=0) and horizontal (axis=1) stacking.
> - `np.stack()` joins arrays along a **new axis** (e.g., stacking three 2D arrays to create a 3D array).
