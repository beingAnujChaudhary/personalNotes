# 📘 Module 1: Plotly Express & Multi-Dimensional Scatter

## 🧠 1. Theory & Fun Fact
**Plotly Express (px)** is a high-level API that lets you create complex, interactive plots with a single line of code. 
💡 **Fun Fact:** Unlike Matplotlib, Plotly plots are rendered as HTML/JavaScript. You can zoom, pan, hover, and toggle legends natively in the browser!

## 💡 2. Mental Model
Think of Plotly Express as a **Swiss Army Knife**. You pass your DataFrame, and map columns to visual aesthetics (`x`, `y`, `color`, `size`, `hover_name`, `animation_frame`).

## 📊 3. Visual Diagram
```text
Standard Scatter:      Multi-Dimensional Scatter:
Y (Life Exp)           Y (Life Exp)
|  *                     |  🔴 (Large, Asia)
| *                      |  🔵 (Small, Europe)
+------- X (GDP)         +------- X (GDP)
                         Size = Population
                         Color = Continent
```

## 💻 4. Code Example: The Ultimate Scatter & Animation


````python
# 1. Multi-Dimensional Scatter (Year 2007)
fig = px.scatter(
    df_gapminder.query("year == 2007"), 
    x="gdpPercap", y="lifeExp", 
    size="pop", color="continent", 
    hover_name="country", 
    log_x=True, size_max=60,
    title="Global Development in 2007"
)
fig.show()

# 2. Animation (Time Travel!)
fig_anim = px.scatter(
    df_gapminder, 
    x="gdpPercap", y="lifeExp", 
    animation_frame="year", animation_group="country",
    size="pop", color="continent", 
    hover_name="country", 
    log_x=True, size_max=55, 
    range_x=[100, 100000], range_y=[25, 90]
)
fig_anim.show()
````

## ⚠️ 5. Common Mistakes
**Forgetting `animation_group`:** If you don't set `animation_group="country"`, Plotly will morph the *order* of the data points instead of tracking the same country across years!

## 🎤 6. Interview Question
**Q:** What is the difference between Plotly Express and Plotly Graph Objects (`go`)?
**A:** Plotly Express (`px`) is a high-level wrapper designed for rapid EDA (one-liners). Plotly Graph Objects (`go`) is the low-level, foundational API used for complex, custom layouts (like 3D surfaces or multi-axis subplots). `px` actually generates `go` figures under the hood!

# 📘 Module 2: Line, Bar & Histograms

## 🧠 1. Theory & Mental Model
- **Line Plots:** Track continuous change (Populations over time).
- **Bar Charts:** Compare discrete categories (Total population by continent).
- **Histograms:** Show frequency distributions (How many countries have a life expectancy between 60-70?).

## 💻 2. Code Examples


````python
# 1. Line Plot (India, China, Pakistan Population)
fig_line = px.line(
    df_gapminder.query("country in ['India', 'China', 'Pakistan']"), 
    x="year", y="pop", color="country", markers=True,
    title="Population Growth: Asia's Giants"
)
fig_line.show()

# 2. Stacked Bar Chart (2007 Pop Contribution)
df_2007 = df_gapminder.query("year == 2007")
fig_bar = px.bar(
    df_2007, x="continent", y="pop", color="country",
    title="2007 Population Contribution by Country (Stacked)"
)
fig_bar.show()

# 3. Histogram with text_auto
fig_hist = px.histogram(
    df_gapminder.query("year == 2007"), 
    x="lifeExp", nbins=30, color="continent",
    text_auto=True, # Shows the count on top of each bar!
    title="Distribution of Life Expectancy in 2007"
)
fig_hist.update_traces(textfont_size=10)
fig_hist.show()
````

## 🎤 3. Interview Question
**Q:** What does the `text_auto=True` parameter do in a Plotly Histogram or Bar chart?
**A:** It automatically calculates the aggregate value (count, sum, etc.) for each bin/bar and renders that number directly on top of the shape, eliminating the need to hover to see the exact frequency.

# 📘 Module 3: Pie, Sunburst & Heatmaps

### 📝 [Markdown Cell]
```markdown
## 🧠 1. Theory & Mental Model
- **Pie Charts:** Show part-to-whole relationships. (Avoid if > 5 categories).
- **Sunburst:** The "Pie Chart on Steroids". Visualizes hierarchical data radially.
- **Heatmaps:** Matrix visualization where color intensity represents magnitude.

## 💡 2. Fun Fact
Plotly Express doesn't have an `explode` parameter like Matplotlib. Instead, you use `fig.update_traces(pull=[0.1, 0, 0])` to "pull" or explode specific slices outward!

## 💻 3. Code Examples


````python
# 1. Pie Chart (Exploded)
df_1952 = df_gapminder.query("year == 1952")
cont_pop = df_1952.groupby("continent")["pop"].sum().reset_index()
fig_pie = px.pie(cont_pop, values="pop", names="continent", title="World Pop 1952")
fig_pie.update_traces(pull=[0.1, 0, 0, 0, 0]) # Explode the first slice!
fig_pie.show()
````

````python
# 2. Sunburst (Hierarchy: Continent -> Country)
fig_sun = px.sunburst(
    df_gapminder.query("year == 2007"), 
    path=['continent', 'country'], 
    values='pop', 
    color='lifeExp', 
    title="Global Population Hierarchy"
)
fig_sun.show()
````

````python
# 3. Heatmap (Continent vs Year -> Avg Life Exp)
heatmap_data = df_gapminder.groupby(['continent', 'year'])['lifeExp'].mean().reset_index()
pivot_data = heatmap_data.pivot(index='continent', columns='year', values='lifeExp')
fig_heat = px.imshow(
    pivot_data, text_auto=".2f", aspect="auto", 
    color_continuous_scale='Viridis',
    title="Average Life Expectancy Heatmap"
)
fig_heat.show()
````

## 🎤 4. Interview Question
**Q:** When would you choose a Sunburst chart over a Treemap?
**A:** Use Sunburst when the **hierarchical path** (parent-to-child relationship) is the most important story you want to tell. Use Treemap when you just want to compare the **relative sizes** of nested categories using rectangular area, which is often easier for the human eye to compare than angles.

# 📘 Module 4: 3D, Facets, Graph Objects & Subplots

## 🧠 1. Theory & Mental Model
When Plotly Express isn't enough, we drop down to **Plotly Graph Objects (`go`)** and **`make_subplots`**.
- **3D Scatter:** Points in $(X, Y, Z)$ space.
- **Scatter Matrix:** A grid of scatter plots showing pairwise relationships.
- **Facet Plots:** Small multiples (splitting one chart into a grid based on a category).
- **Surface/Contour:** 3D meshes and topographic maps (Requires `go`).

## 💻 2. Code Examples


````python
# 1. 3D Scatter Plot (Express)
df_2007 = df_gapminder.query("year == 2007")
fig_3d = px.scatter_3d(
    df_2007, x='gdpPercap', y='lifeExp', z='pop',
    color='continent', hover_name='country',
    title="3D Global Development"
)
fig_3d.show()
````

````python
# 2. Scatter Matrix (Iris Dataset)
fig_scatmat = px.scatter_matrix(
    df_iris, 
    dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], 
    color="species",
    title="Iris Feature Correlations"
)
fig_scatmat.show()
````

````python
# 3. Facet Plot (Small Multiples)
fig_facet = px.scatter(
    df_gapminder.query("year == 2007"), 
    x="gdpPercap", y="lifeExp", color="continent", 
    facet_col="continent",
    title="GDP vs Life Exp by Continent"
)
fig_facet.show()
````

````python
# 4. 3D Surface & Contour (Graph Objects)
z_data = np.random.rand(50, 50)

fig_surface = go.Figure(data=[go.Surface(z=z_data, colorscale='Viridis')])
fig_surface.show()
````

````python
fig_contour = go.Figure(data=go.Contour(z=z_data, colorscale='Viridis'))
fig_contour.show()
````

````python
# 5. Subplots (Combining different chart types)
fig_sub = make_subplots(rows=1, cols=2, specs=[[{"type": "scatter"}, {"type": "bar"}]])
fig_sub.add_trace(go.Scatter(x=[1, 2, 3], y=[4, 5, 6]), row=1, col=1)
fig_sub.add_trace(go.Bar(x=[1, 2, 3], y=[10, 8, 12]), row=1, col=2)
fig_sub.show()
````

## ⚠️ 3. Common Mistakes
**Mixing `px` and `go` incorrectly:** You can't just pass a `px` figure into a `go` subplot. You must extract the traces: `fig_sub.add_trace(fig_px.data[0], row=1, col=1)`.

## 🎤 4. Interview Question
**Q:** How do you combine a Plotly Express chart and a Plotly Graph Object chart into the same dashboard?
**A:** You use `make_subplots()`. You create the Express chart, then iterate through its `fig.data` (which contains the traces) and add them to the subplot grid using `add_trace(trace, row=r, col=c)`.
