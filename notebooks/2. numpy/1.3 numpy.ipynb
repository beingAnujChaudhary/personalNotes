# 🎯 Concept 2: Conditional Logic & Searching

## 📝 Theory
- **`np.where(condition, [x, y])`**: The vectorized version of `if-else`. If `x` and `y` are provided, it returns elements from `x` where condition is True, else `y`. If only `condition` is provided, it returns the **indices** where True.
- **`np.argmax()` / `np.argmin()`**: Returns the *indices* of the maximum/minimum values.
- **`np.isin(element, test_elements)`**: Checks if each element of an array is present in a test list.

## 📊 Visual Diagram
```text
Array:      [ 1, -2,  3, -4,  5]
Condition:  [ T,  F,  T,  F,  T]  (arr > 0)

np.where(cond, arr, 0):
Result:     [ 1,  0,  3,  0,  5]  (Replaces negatives with 0 -> ReLU!)

np.where(cond):
Result:     (array([0, 2, 4]),)   (Returns indices of True)
```

## ⚠️ Common Mistake
Forgetting that `np.where(condition)` returns a **tuple** of arrays (one for each dimension). To get a flat array of indices, use `np.where(condition)[0]` or `np.nonzero(condition)[0]`.



### `np.where`

````python
# Sample array for context
arr1D = np.array([2, 8, 4, 9, 5, 12])
print(arr1D)

# ---------------------------------------------------------
# Usage 1: Finding Indices (The "Where is it?" method)
# ---------------------------------------------------------
# When you only provide a condition (no replacement values), 
# np.where() returns the *index positions* where the condition is True.
# In our array [2, 8, 4, 9, 5, 12], the values > 5 are 8, 9, and 12.
# These are located at index 1, 3, and 5.
# Output: (array([1, 3, 5]),)
np.where(arr1D > 5)
````

````python
# ---------------------------------------------------------
# Usage 2: Conditional Replacement (The "If This, Then That" method)
# ---------------------------------------------------------
# Syntax: np.where(condition, value_if_true, value_if_false)
# This scans every element in the array. 
# If an element is > 5, it gets replaced by 10.
# If it is NOT > 5 (meaning it's 5 or less), it stays as its original value (arr1D).
# Output: [ 2 10  4 10  5 10]
np.where(arr1D > 5, 10, arr1D)
````

````python
# Capping extreme values (Outlier Control): Imagine you are processing test scores out of 100, but some glitch caused a few scores to be recorded as 105.

scores = np.array([85, 92, 105, 78, 110, 95])

# "If score > 100, make it 100. Otherwise, leave it alone."
clean_scores = np.where(scores > 100, 100, scores)

print(clean_scores)
# Output: [ 85  92 100  78 100  95]
````

````python
# Binary Classification: You have an array of model predictions (probabilities between 0 and 1) and you need to convert them into 1 (True) or 0 (False).

probabilities = np.array([0.1, 0.8, 0.4, 0.9, 0.51])

# "If prob >= 0.5, it's a 1. Otherwise, it's a 0."
predictions = np.where(probabilities >= 0.5, 1, 0)

print(predictions)
# Output: [0 1 0 1 1]
````

````python
# Replacing Specific Values (Handling Errors): Sometimes datasets use -1 to represent missing data. You want to replace them with 0.
data = np.array([45, 62, -1, 55, -1, 80])

# "If data equals -1, change to 0. Otherwise, keep data."
fixed_data = np.where(data == -1, 0, data)

print(fixed_data)
# Output: [45 62  0 55  0 80]
````

### `np.argmax`

The most common mistake beginners make is thinking argmax() returns the actual maximum number.
- `np.max()` asks: "What is the highest score?"
- `np.argmax()` asks: "Who got the highest score?" (It points to the location).

To actually get the highest value using `argmax`, you would pass that index back into the array: `highest_value = a[np.argmax(a)]`.

````python
a = np.array([33, 75, 18, 14, 69, 10, 69, 39, 98, 7])
print(a)

# ---------------------------------------------------------
# 1D Array argmax/argmin
# ---------------------------------------------------------
# np.argmax() returns the INDEX of the highest value, not the value itself.
# In our example, the highest value is 98, which is at index 8.
print(np.argmax(a)) 

# np.argmin() returns the INDEX of the lowest value.
# The lowest value is 7, which is at index 9.
print(np.argmin(a)) 
````

````python
# ---------------------------------------------------------
# 2D Array argmax/argmin (Assuming the arr2D shown above)
# ---------------------------------------------------------

# axis=0 means "look down the columns" (compare rows).
print(np.argmax(arr2D, axis=0)) 

# axis=1 means "look across the rows" (compare columns).
print(np.argmax(arr2D, axis=1)) 
````

````python
# Same logic applies for argmin, but finding the lowest values.
# Look down columns: 
print(np.argmin(arr2D, axis=0)) 

# Look across rows: 
print(np.argmin(arr2D, axis=1)) 
````

````python
# Machine Learning Predictions (Classification)
# When a neural network makes a prediction, it usually outputs an array of probabilities for each category. argmax is the standard way to find out which category the model is most confident about.

# Imagine a model predicting if an image is a [Cat, Dog, Bird]
# It outputs these probabilities:
probabilities = np.array([0.15, 0.80, 0.05])

categories = ["Cat", "Dog", "Bird"]

# Find the index of the highest probability
predicted_index = np.argmax(probabilities)

print(f"The model predicts: {categories[predicted_index]}")
# Output: The model predicts: Dog
````

````python
# Data Analysis (Finding the "When" or "Where")
# If you have time-series data, you often want to know when the peak or valley occurred, not just what the value was.

# Monthly sales data for a year (Jan to Dec)
sales = np.array([120, 150, 110, 200, 250, 300, 280, 210, 180, 400, 350, 390])
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Find out WHICH month had the worst sales
worst_month_index = np.argmin(sales)

print(f"Worst sales were in {months[worst_month_index]} with {sales[worst_month_index]} units.")
# Output: Worst sales were in Mar with 110 units.
````

### `np.isin`

`np.isin` is one of the most powerful filtering tools in NumPy. Instead of writing nested `for loops` to check if items in one massive list exist in another massive list, np.isin() handles it instantly using vectorized C-code under the hood.

The resulting [False, True, False...] array is called a **Boolean Mask**. You overlay this mask onto any array of the same shape to filter it.

````python
# Let's define 'a' as an array of the exact same length as 'b' (5 elements).
# Imagine 'a' is a list of customer names, and 'b' is their ID numbers.
a = np.array(["Alice", "Bob", "Charlie", "David", "Eve"])
b = np.array([10, 20, 30, 40, 50])

# This is the list we want to check against (e.g., a list of VIP IDs)
c = np.array([20, 40, 60])

# ---------------------------------------------------------
# Step 1: The Boolean Mask
# ---------------------------------------------------------
# np.isin() goes through every single element in 'b' and asks: 
# "Is this number anywhere inside 'c'?"
# It returns an array of True/False values matching the length of 'b'.
# 10 (No), 20 (Yes), 30 (No), 40 (Yes), 50 (No)
mask = np.isin(b, c)

print(mask) 
# Output: [False  True False  True False]
````

````python
# ---------------------------------------------------------
# Step 2: Boolean Indexing (Filtering)
# ---------------------------------------------------------
# By placing the True/False mask inside the brackets of 'a', 
# NumPy acts like a filter. It only keeps the items in 'a' where the mask is True.
# Since the mask is True at index 1 and index 3, it grabs "Bob" and "David".
filtered_a = a[mask]

print(filtered_a)
# Output: ['Bob' 'David']
````

````python
# Filtering a dataset based on a specific target list
# Imagine you are cleaning a massive Kaggle dataset or pulling reservation data from a flight database. You have an array of millions of transactions, but you only want to analyze the ones that belong to three specific regions.

# Array of 100,000 transaction region codes
transaction_regions = np.array(["EU", "US", "IN", "UK", "IN", "EU", "AU"]) # ... shortened for example

# We only want to analyze transactions from India, the US, and the UK
target_regions = ["IN", "US", "UK"]

# Create the mask and filter immediately
valid_transactions = transaction_regions[np.isin(transaction_regions, target_regions)]

print(valid_transactions)
# Output: ['US' 'IN' 'UK' 'IN']
````

````python
# Dropping unwanted data (The ~ NOT operator)
# You can easily reverse the behavior of np.isin() by putting a tilde (~) in front of it. This flips all the True values to False and vice versa, which is perfect for removing anomalies or blacklisted items.

product_ids = np.array([101, 102, 103, 104, 105])
discontinued_products = [102, 105]

# The '~' means "NOT in". Keep the products that are NOT discontinued.
active_products = product_ids[~np.isin(product_ids, discontinued_products)]

print(active_products)
# Output: [101 103 104]
````

## ⚠️ Common Mistake
Using `np.append()` inside a `for` loop to build an array. 
**Why?** NumPy arrays are fixed in memory. `append` creates a brand new array and copies all data every single time, resulting in **O(N²)** time complexity! Always pre-allocate or use Python lists for loops, then convert to NumPy.



### `np.concatenate`

````python
# axis=0 means "concatenate along the rows" (stacking vertically). 
# This places the original arr2D on top, and stacks the sorted version directly underneath it.
# If arr2D had 3 rows and 4 columns, 'result' will now have 6 rows and 4 columns.
result = np.concatenate((arr2D, np.sort(arr2D)), axis=0)
print(result)
````

````python
# axis=1 means "concatenate along the columns" (stacking horizontally).
# This places the original arr2D on the left, and glues the sorted version to its right side.
# If arr2D had 3 rows and 4 columns, 'result' will now have 3 rows and 8 columns.
result = np.concatenate((arr2D, np.sort(arr2D)), axis=1)
print(result)
````

````python
# axis=0: Concatenating along the blocks (the "pages"). 
# This takes the "pages" from the sorted 3D array and adds them to the back of the original array.
# The number of rows and columns stays the same, but the total number of blocks doubles.
stacked_vertically = np.concatenate((arr3D, np.sort(arr3D)), axis=0)
print(stacked_vertically)
````

````python
# axis=1: Concatenating along the rows (on each individual page).
# For every single block/page, this takes the rows from the sorted version and 
# adds them to the bottom of the original rows. The number of blocks stays the same, 
# but each block now has twice as many rows.
stacked_horizontally = np.concatenate((arr3D, np.sort(arr3D)), axis=1)
print(stacked_horizontally)
````

````python
# axis=2: Concatenating along the columns (on each individual page).
# For every block/page, this takes the columns from the sorted version and 
# glues them to the right side of the original columns. Each block now has twice as many columns.
stacked_depth = np.concatenate((arr3D, np.sort(arr3D)), axis=2)
print(stacked_depth)
````

### `np.append`
When appending arrays together along a specific axis, their dimensions must match on the other axes. By using arr2D.shape[0], this code cleverly ensures that the newly generated column has the exact same number of rows as the original matrix, preventing a common NumPy shape mismatch error.


````python
# Examples of append

# 1. Appending to a 1D array
# This adds the value 200 to the end of 'arr1D'. 
# Note: np.append does NOT modify the original array in place; it returns a new array.
# To keep the result, you would need to assign it to a variable, e.g., arr1D = np.append(arr1D, 200)
np.append(arr1D, 200)
````

````python
# 2. Appending a column of 1s to a 2D array
print(np.append(
    arr2D,                               # The original 2D array
    np.ones((arr2D.shape[0], 1)),        # Creates a new 2D array of 1s. 
                                         # 'arr2D.shape[0]' ensures it has the exact same number of rows as arr2D, and '1' gives it a single column.
    axis=1                               # axis=1 tells NumPy to append horizontally (adding a new column). axis=0 would append a new row.
))
````

````python
# 3. Appending two columns of random numbers to a 2D array
print(np.append(
    arr2D,                               # The original 2D array
    np.random.random((arr2D.shape[0], 2)), # Creates a new 2D array filled with random floats between 0.0 and 1.0. 
                                         # Again, the shape matches the number of rows of arr2D and adds 2 columns.
    axis=1                               # Appends this new random column horizontally to the right side of arr2D.
))
````

````python

# 1. Create the original 3D array using randint
# Let's imagine this as 2 separate blocks, where each block is a 3x4 grid.
# Shape: (2, 3, 4) -> (blocks/depth, rows, columns)
arr3D = np.random.randint(1, 10, size=(2, 3, 4))

# --- THE CRITICAL STEP ---
# We upgrade the 2D array to 3D. 
# np.expand_dims adds a new axis. By choosing axis=0, we change the 
# shape from (3, 4) to (1, 3, 4). It is now technically a 3D array containing one block.
arr2D_expanded = np.expand_dims(arr2D, axis=0)

# 3. Now we can safely append them together
result = np.append(
    arr3D,           # The original 3D array
    arr2D_expanded,  # The 2D array we just converted to 3D
    axis=0           # axis=0 means we are adding a new 'block' to the depth stack
)

print(f"Original 3D shape: {arr3D.shape}")
print(f"Original 2D shape: {arr2D.shape}")
print(f"Expanded 2D shape: {arr2D_expanded.shape}")
print(f"Final Appended shape: {result.shape}")
````

### `np.expand_dims`

````python

#  Expanding along axis=0 (Adding a "Row" dimension)
# axis=0 inserts a new dimension at the very beginning (index 0).
# This effectively wraps your entire 1D array inside a new set of brackets, 
# turning it into a 2D "row vector".
print(np.expand_dims(arr1D, axis=0), np.expand_dims(arr1D, axis=0).shape)
````

````python
# Expanding along axis=1 (Adding a "Column" dimension)
# axis=1 inserts a new dimension at the second position (index 1).
# This wraps *each individual element* in its own set of brackets, 
# turning it into a 2D "column vector".
print(np.expand_dims(arr1D, axis=1), np.expand_dims(arr1D, axis=1).shape)
````

> `np.expand_dims`
- **What it does**: It only does one thing: it inserts a "dummy" dimension (a dimension of length 1) at a specific axis.
- **Why/When to use it**: Use this when you want to increase the dimensionality of an array without changing its fundamental structure, and especially when you don't know (or **don't want to calculate) the original shape**.
- **Common Use Case**: In machine learning, models often expect data in batches. If you have a single image shaped (`28, 28`), you can use `np.expand_dims(image, axis=0)` to instantly turn it into (1, 28, 28)—a batch of one image—without needing to manually type `28, 28`.
> `np.reshape`
- **What it does**: It completely restructures the data into a brand new shape, as long as the total number of elements remains the same.
- **Why/When to use it**: Use this when you are fundamentally reorganizing your data, such as turning a long 1D list into a 2D matrix, or flattening a 3D array into 1D.
- Common Use Case: If you have 12 data points in a 1D array and you want to organize them into a calendar view of 3 weeks (rows) with 4 days (columns).
*Note: You can use `np.reshape(arr1D, (1, -1))` to mimic what `expand_dims` does, but `expand_dims` is considered much more readable and explicit when your specific goal is just to add a single dimension*.

> Practical Examples of Both
Using `np.expand_dims` to prep data for a machine learning model:


````python
# A single color image (Height, Width, Color Channels)
single_image = np.random.randint(0, 255, size=(256, 256, 3))

# Neural networks usually expect (Batch Size, Height, Width, Channels)
# We easily add a batch dimension to the front
batch_ready_image = np.expand_dims(single_image, axis=0)

print(batch_ready_image.shape) 
````

````python
# A flat array of 12 numbers
flat_data = np.arange(12) 

# Reorganize it into a 3x4 matrix
matrix_data = np.reshape(flat_data, (3, 4))

print(matrix_data)
````

### `np.swapaxes`

`np.swapaxes()` interchanges two specified axes of an array without copying data (it returns a view when possible). This is crucial for reshaping data to match algorithm expectations.

- **Does NOT create a copy by default** (returns a view when strides allow)
- **Key difference from transpose**: `transpose()` reverses ALL axes; `swapaxes()` swaps only two
- **Common in image processing**: Changing from (height, width, channels) to (channels, height, width)


````python
# Sample 3D array (2 blocks, 3 rows, 4 columns)
arr3D = np.arange(24).reshape(2, 3, 4)
print("Original shape:", arr3D.shape)  # (2, 3, 4)
print("Original:\n", arr3D)

# ---------------------------------------------------------
# np.swapaxes: Swap any two axes of an array
# ---------------------------------------------------------
# Syntax: np.swapaxes(array, axis1, axis2)
# Swaps axis 0 (depth) with axis 1 (rows)
# Shape changes from (2, 3, 4) to (3, 2, 4)
swapped = np.swapaxes(arr3D, 0, 1)

print("\nSwapped shape:", swapped.shape)  # (3, 2, 4)
print("Swapped (axes 0 and 1):\n", swapped)

# Swapping axes 1 and 2: (2, 3, 4) -> (2, 4, 3)
swapped2 = np.swapaxes(arr3D, 1, 2)
print("\nSwapped shape (axes 1,2):", swapped2.shape)  # (2, 4, 3)
````

### `np.cumsum` & `cumprod`

- `np.cumsum` (Cumulative Sum): Think of this as a running total. Instead of just giving you the final sum, it gives you an array showing the sum at every single step.
- `np.cumprod` (Cumulative Product): The exact same concept as cumsum, but it multiplies the numbers instead of adding them.
- `np.prod` (Product): This multiplies all elements together but only returns the final result, not the step-by-step array.

*Note: If you run `np.cumsum(arr2D)` without specifying an axis, NumPy will flatten your 2D array into a 1D line first, and then calculate the running total.*

````python
# ---------------------------------------------------------
# 1D Array: Cumulative Sum and Total Product
# ---------------------------------------------------------
# np.cumsum(): Keeps a "running total" as it moves through the array.
# np.prod(): Multiplies EVERY number in the array together into a single value.
print(arr1D, np.cumsum(arr1D), np.prod(arr1D), sep="\n")
````

````python
# ---------------------------------------------------------
# 2D Array: axis=0 (Looking Down the Columns)
# ---------------------------------------------------------
# axis=0 applies the logic vertically (column by column).
# np.cumsum(arr2D, axis=0): Running total down the columns.
# np.cumprod(arr2D, axis=0): Running multiplication down the columns.
print(arr2D, np.cumsum(arr2D, axis=0), np.cumprod(arr2D, axis=0), sep="\n")
````

````python
# ---------------------------------------------------------
# 2D Array: axis=1 (Looking Across the Rows)
# ---------------------------------------------------------
# axis=1 applies the logic horizontally (row by row).
# np.cumsum(arr2D, axis=1): Running total across each row.
# np.cumprod(arr2D, axis=1): Running multiplication across each row.
print(arr2D, np.cumsum(arr2D, axis=1), np.cumprod(arr2D, axis=1), sep="\n")
````

````python
# Finance (Running Totals / Year-to-Date)
# If you have an array representing monthly profits, cumsum instantly gives you your Year-to-Date (YTD) total at each month.

# Profits for Jan, Feb, Mar, Apr
monthly_profits = np.array([1000, 1500, -500, 2000])

# Calculate Year-to-Date total at each step
ytd_profits = np.cumsum(monthly_profits)

print(ytd_profits)
# Output: [1000 2500 2000 4000]
# By April, the total profit was 4000.
````

````python
# Statistics (Cumulative Distribution Function - CDF)
# In probability, if you have an array of percentages representing how likely an event is to happen at different stages, cumsum helps you find the probability that the event will happen by a certain stage.

# Daily probability of a machine breaking down (Day 1, Day 2, Day 3)
daily_prob = np.array([0.10, 0.15, 0.05])

# Probability it has broken down BY that day
cumulative_prob = np.cumsum(daily_prob)

print(cumulative_prob)
# Output: [0.10 0.25 0.30]
# There is a 25% chance it breaks down by the end of Day 2.
````

````python
# For Compound Interest or Investment Growth
# If you have an array representing the daily or yearly growth multiplier of a stock, cumprod shows you the total value of your investment over time.

# A stock grows by 5% (1.05), drops by 2% (0.98), grows by 10% (1.10)
daily_multipliers = np.array([1.05, 0.98, 1.10])
initial_investment = 1000

# Calculate the growth journey
value_over_time = initial_investment * np.cumprod(daily_multipliers)

print(value_over_time)
# Output: [1050.   1029.   1131.9]
````

### `np.histogram` & `np.corrcoef`

The biggest misconception is that `np.histogram` function plots a chart. It does not. It takes a messy array of continuous numbers and sorts them into discrete buckets (bins). It returns two arrays:
1. **Counts**: The number of items that fell into each bucket.
2. **Bin Edges**: The boundaries of those buckets.

`np.corrcoef` measures the linear relationship between two variables using the Pearson correlation formula. The resulting number always falls between -1.0 and 1.0:
1. **1.0**: Perfect positive correlation (As X goes up, Y goes up identically).
2. **0.0**: Absolutely no relationship (Random scatter).
3. **-1.0**: Perfect negative correlation (As X goes up, Y goes down identically).

*Note on the Matrix Output: When you pass two arrays to `np.corrcoef(x, y)`, it returns a 2x2 grid. The top-left and bottom-right will always be `1.0` (because X correlates perfectly with itself, and Y correlates perfectly with itself). The other two numbers are the actual relationship between X and Y.*

````python
# ==========================================
# np.histogram: Counting data into "bins"
# ==========================================
# Let's say we have the ages of 10 people in a restaurant
ages = np.array([22, 25, 29, 35, 42, 45, 48, 55, 62, 70])

# np.histogram DOES NOT draw a picture. It does the math behind the picture.
# We tell it to group the ages into 3 equal-width "bins" (e.g., young, middle, older).
counts, bin_edges = np.histogram(ages, bins=3)

# 'counts' tells us HOW MANY people are in each bin.
print(f"Counts: {counts}") 
# 'bin_edges' tells us the start and end numbers for each bin.
# Note: There is always 1 more bin_edge than there are counts.
print(f"Bin Edges: {bin_edges}\n")
````

````python
# ==========================================
# np.corrcoef: Finding linear relationships
# ==========================================
# Let's look at the relationship between hours studied and test scores
hours_studied = np.array([1, 2, 3, 4, 5, 6])
test_scores =   np.array([50, 55, 65, 70, 85, 90])

# np.corrcoef returns a "Correlation Matrix".
correlation_matrix = np.corrcoef(hours_studied, test_scores)

print("Correlation Matrix:")
print(correlation_matrix)

# To get just the specific correlation number between the two variables, 
# you extract the number from row 0, column 1 (or row 1, column 0)
actual_correlation = correlation_matrix[0, 1]
print(f"\nThe correlation coefficient is: {actual_correlation:.3f}")
````

````python
# For Image Processing (Exposure Check)
# Here, we simulate an "overexposed" image (mostly bright pixels) and use a histogram to detect that the highlights are blown out.

# 1. Simulate an overexposed 1080p image (1080 x 1920 pixels)
# We generate random pixel values, but heavily skew them towards 255 (white)
image_pixels = np.random.randint(200, 256, size=(1080, 1920))

# 2. Define our brightness buckets (bins)
# Bin 1: 0-85 (Shadows/Dark)
# Bin 2: 85-170 (Midtones)
# Bin 3: 170-255 (Highlights/Bright)
bins = [0, 85, 170, 255]

# 3. Run the histogram
counts, bin_edges = np.histogram(image_pixels, bins=bins)

shadows, midtones, highlights = counts

print(f"Shadow pixels: {shadows}")
print(f"Midtone pixels: {midtones}")
print(f"Highlight pixels: {highlights}")

# 4. The Algorithm Decision
total_pixels = image_pixels.size
if highlights / total_pixels > 0.8:
    print("\nSYSTEM ALERT: Image is overexposed! Triggering auto-darken algorithm...")
````

# 🧮 Concept 5: Set Operations, Clipping & Comparisons

## 📝 Theory
- **Set Operations**: `np.union1d`, `np.intersect1d`, `np.setdiff1d`, `np.setxor1d`. (Note: These flatten arrays and return **unique, sorted** results).
- **`np.clip(arr, min, max)`**: Limits values to a specific range. Values below `min` become `min`, above `max` become `max`.
- **`np.allclose(a, b)`**: Checks if two arrays are element-wise equal within a tolerance (crucial for floats!).
- **`np.array_equal(a, b)`**: Checks if two arrays have the same shape and exact elements.
- **`np.random.uniform(low, high, size)`**: Draws samples from a uniform distribution.

## 📊 Visual Diagram
```text
A = [1, 2, 3]
B = [3, 4, 5]

intersect1d: [3]
union1d:     [1, 2, 3, 4, 5]
setdiff1d(A,B): [1, 2] (In A but not B)

Clipping:
Input:  [-5, 0, 5, 10]
Clip(0, 5): [0, 0, 5, 5]
```

## ⚠️ Common Mistake
Using `==` to compare floating-point NumPy arrays. Due to precision issues (e.g., `0.1 + 0.2 != 0.3`), `==` will return `False`. **Always use `np.allclose()` for floats.**

````python
# Sample arrays
setA = np.array([1, 2, 3, 4, 5])
setB = np.array([4, 5, 6, 7, 8])

# ---------------------------------------------------------
# 1. Union (Combine everything, remove duplicates)
# ---------------------------------------------------------
# Like a SQL FULL OUTER JOIN or UNION. It takes all items from both arrays.
# It automatically sorts the result and drops any duplicate numbers.
print(np.union1d(setA, setB))


# ---------------------------------------------------------
# 2. Intersection (Find the common ground)
# ---------------------------------------------------------
# Like a SQL INNER JOIN. It only returns the numbers that exist in BOTH arrays.
# The numbers 4 and 5 are the only ones present in both setA and setB.
print(np.intersect1d(setA, setB))


# ---------------------------------------------------------
# 3. Set Difference (Find what is missing)
# ---------------------------------------------------------
# Like a SQL EXCEPT or LEFT JOIN (where right is null). 
# It asks: "What is in setA that is NOT in setB?"
# It removes the 4 and 5 because they exist in setB. Order matters here!
print(np.setdiff1d(setA, setB))


# ---------------------------------------------------------
# 4. Set XOR / Symmetric Difference (Find the exclusives)
# ---------------------------------------------------------
# It asks: "What is in ONE of the arrays, but NOT both?"
# It drops the shared numbers (4 and 5) and keeps the unique ones from both sides.
print(np.setxor1d(setA, setB))


````

````python
# To find overlapping customers
# You want to run a targeted marketing campaign for high-value customers who buy both electronics and clothing.

electronics_buyers = np.array([101, 105, 202, 305, 400])
clothing_buyers = np.array([202, 300, 400, 500])

# Find customers who bought from BOTH categories
cross_shoppers = np.intersect1d(electronics_buyers, clothing_buyers)

print(f"Send promo codes to Customer IDs: {cross_shoppers}")
# Output: Send promo codes to Customer IDs: [202 400]
````

### `np.clip(array, min, max)`

`np.clip()` is a fast, vectorized way to apply both a floor (minimum limit) and a ceiling (maximum limit) to your data simultaneously.
1. You don't need both: If you only want to set a floor but don't care about a ceiling, you can pass `None` as the max value: `np.clip(arr1D, 3, None)`.
2. It creates a copy: Like most NumPy functions, it returns a brand new array and leaves the original untouched.

*Note: You could achieve the exact same result using `np.where(arr1D < 3, 3, np.where(arr1D > 6, 6, arr1D))`, but `np.clip` is vastly cleaner and faster.*

````python
arr1D = np.array([1, 2, 3, 4, 5, 6, 7, 8])
print("Original:", arr1D)

# ---------------------------------------------------------
# np.clip: Setting boundaries (Floors and Ceilings)
# ---------------------------------------------------------
# Syntax: np.clip(array, min_value, max_value)
# It forces all numbers in the array to stay within the range [3, 6].
# - Any number smaller than 3 is bumped UP to 3.
# - Any number larger than 6 is pulled DOWN to 6.
# - Numbers already between 3 and 6 are left alone.
clipped_arr = np.clip(arr1D, 3, 6)

print("Clipped: ", clipped_arr)
````

````python
# Image Processing (Pixel Boundaries)
# As we've established, images are arrays of pixel values that must stay between 0 (pitch black) and 255 (pure white). If you apply a brightening algorithm that adds 50 to every pixel, a pixel that was already at 220 will become 270. This will crash image software.

# A slightly bright image
pixels = np.array([150, 200, 220, 240])

# Brighten the image by 50
brightened = pixels + 50 
# Output: [200, 250, 270, 290] -> 270 and 290 are illegal pixel values!

# Clip it to legally valid pixel ranges
safe_pixels = np.clip(brightened, 0, 255)

print(safe_pixels)
# Output: [200 250 255 255]
````

````python
# Machine Learning (Gradient Clipping)
# When training deep neural networks, there is a common bug called "Exploding Gradients," where the mathematical updates to the model become so massively large that the model breaks and outputs NaN (Not a Number). Data scientists use np.clip to force the updates to stay within a safe range (like -1.0 to 1.0).

# Raw mathematical updates for the AI
gradients = np.array([-0.5, 0.2, -99.9, 0.8, 150.5]) 
# -99.9 and 150.5 are massive spikes that will destroy the model.

# Clip the gradients to keep the learning process stable
safe_gradients = np.clip(gradients, -1.0, 1.0)

print(safe_gradients)
# Output: [-0.5  0.2 -1.   0.8  1. ]
````

````python
# Sensor Data (Filtering Outlier Spikes)
# If you have a temperature sensor that normally reads between 60°F and 80°F, but occasionally glitches and outputs 999°F, clipping keeps your charts readable.

# Hourly temperature readings
temps = np.array([68, 70, 72, 999, 71, 69])

# Clip out the impossible hardware glitches (assume max logical temp is 120)
clean_temps = np.clip(temps, 0, 120)

print(clean_temps)
# Output: [ 68  70  72 120  71  69]
````

```python
count = np.count_nonzero(arr)
```

- **Memory efficient**: Doesn't create intermediate arrays when used with conditions

## 6. np.allclose

### Code Explanation

```python
import numpy as np

# ---------------------------------------------------------
# np.allclose: Check if arrays are element-wise equal within tolerance
# ---------------------------------------------------------
# Syntax: np.allclose(a, b, rtol=1e-05, atol=1e-08)
# Returns True if |a - b| <= atol + rtol * |b| for all elements

# Exact match
a = np.array([1.0, 2.0, 3.0])
b = np.array([1.0, 2.0, 3.0])
print("Exact match:", np.allclose(a, b))  # True

# Floating point differences
x = np.array([0.1 + 0.2])  # 0.30000000000000004
y = np.array([0.3])         # 0.3
print("0.1+0.2 == 0.3:", x == y)           # False (floating point issue!)
print("allclose(0.1+0.2, 0.3):", np.allclose(x, y))  # True

# Custom tolerances
arr1 = np.array([1.0, 2.0, 3.0])
arr2 = np.array([1.01, 1.99, 3.02])
print("Default tolerance:", np.allclose(arr1, arr2))  # False
print("Relaxed tolerance:", np.allclose(arr1, arr2, atol=0.1))  # True

# NaN handling
arr_nan1 = np.array([1.0, np.nan, 3.0])
arr_nan2 = np.array([1.0, np.nan, 3.0])
print("NaN comparison:", np.allclose(arr_nan1, arr_nan2))  # False
print("NaN with equal_nan:", np.allclose(arr_nan1, arr_nan2, equal_nan=True))  # True
```

### The Core Concept

`np.allclose()` compares two arrays element-wise and returns `True` if all elements are equal within given tolerances. The comparison formula is: `|a - b| <= atol + rtol * |b|`

- **rtol (relative tolerance)**: Accounts for magnitude differences (default: 1e-5)
- **atol (absolute tolerance)**: Absolute minimum tolerance (default: 1e-8)
- **Essential for floating-point**: Never use `==` for float arrays

### Practical Use Cases

**Use Case 1: Verifying Numerical Algorithms**
```python

def analytical_gradient(x):

x = np.array([1.0, 2.0, 3.0])
num_grad = numerical_gradient(lambda x: x**2, x)
anal_grad = analytical_gradient(x)
print("Gradient check:", np.allclose(num_grad, anal_grad, rtol=1e-4))
```

**Use Case 2: Unit Testing Machine Learning Models**
```python

actual_output = np.array([0.7000001, 0.1999999, 0.1000002])
print("Output matches:", np.allclose(expected_output, actual_output))
```

**Use Case 3: Checking Matrix Inversion**

## 7. np.array_equal

### Code Explanation

```python
import numpy as np

# ---------------------------------------------------------
# np.array_equal: Check exact equality (shape AND values)
# ---------------------------------------------------------
# Syntax: np.array_equal(a, b)
# Returns True only if arrays have same shape and all elements equal

# Exact match
a = np.array([1, 2, 3])
b = np.array([1, 2, 3])
print("Same array:", np.array_equal(a, b))  # True

# Different shape
c = np.array([[1, 2, 3]])
print("Different shape:", np.array_equal(a, c))  # False

# Different values
d = np.array([1, 2, 4])
print("Different values:", np.array_equal(a, d))  # False

# Integer arrays - exact comparison
int_arr1 = np.array([1, 2, 3])
int_arr2 = np.array([1, 2, 3])
print("Integer equality:", np.array_equal(int_arr1, int_arr2))  # True

# Float comparison issue
float_arr1 = np.array([0.1 + 0.2])
float_arr2 = np.array([0.3])
print("Float equality:", np.array_equal(float_arr1, float_arr2))  # False!
print("Should use allclose for floats")

# NaN comparison
nan_arr1 = np.array([1, np.nan])
nan_arr2 = np.array([1, np.nan])
```

### The Core Concept

`np.array_equal()` returns `True` only if both arrays have **identical shapes** and **all corresponding elements are exactly equal** (bit-for-bit match for integers, exact float representation).

- **Strict comparison**: Even tiny floating-point differences matter
- **Shape-sensitive**: Different shapes always return `False`
- **NOT for floating-point**: Use `np.allclose()` for float comparisons
- **Fast for integers**: Perfect for comparing integer labels, indices, masks

### Practical Use Cases

**Use Case 1: Comparing Label Arrays**

## Comparison: allclose vs array_equal

```python
# When to use which?
import numpy as np

# Integer arrays: USE array_equal
labels1 = np.array([1, 2, 3])
labels2 = np.array([1, 2, 3])
print("Integer comparison:", np.array_equal(labels1, labels2))  # True

# Float arrays: USE allclose
values1 = np.array([0.1, 0.2, 0.3])
values2 = np.array([0.1, 0.2, 0.3])
print("Float with equal:", np.array_equal(values1, values2))  # Could be False!
print("Float with allclose:", np.allclose(values1, values2))  # True (safe)

# Shapes matter: allclose checks shapes implicitly
a = np.array([1.0, 2.0, 3.0])
b = np.array([[1.0, 2.0, 3.0]])
print("Different shapes - allclose:", np.allclose(a, b))  # False
print("Different shapes - equal:", np.array_equal(a, b))  # False
```

## Quick Decision Guide

| Function | Best For | Avoid For |
|----------|----------|-----------|
| **swapaxes** | Reshaping data for different libraries | Simple transposition (use `.T`) |
| **uniform** | Random initialization, sampling | Reproducible sequences without seed |
| **count_nonzero** | Sparsity, completeness, accuracy metrics | Counting specific values (use `np.sum(arr==val)`) |
| **tile** | Repeating entire arrays in patterns | Element-wise repetition (use `repeat`) |
| **repeat** | Upsampling, data augmentation | Array-wise repetition (use `tile`) |
| **allclose** | Floating-point comparisons, testing | Integer/boolean comparisons |
| **array_equal** | Integer/label comparison, exact matching | Float comparisons (use `allclose`) |
