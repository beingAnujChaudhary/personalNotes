# 📘 Type Casting & Value Bounds
**Methods Covered:** `astype()`, `between()`, `clip()`


## 🧠 Data often arrives in the wrong format or with extreme outliers.
- **`astype()`**: Converts a Series from one data type to another.
- **`between(a, b)`**: Checks if values fall within a specific range (returns boolean).
- **`clip(lower, upper)`**: Caps values at a specified minimum and maximum threshold.

## 🎯 Why Do We Need It?
Machine Learning models will crash if numeric data is stored as strings. Furthermore, impossible outliers (like a human blood pressure of 400) will completely destroy statistical models. We must cast types and enforce logical bounds.

## 💡 Mental Model
- **`astype`** is a **Translator** (changing the language of the data).
- **`clip`** is a **Bouncer** at a club (enforcing strict min/max limits. Too high? Get pushed down to the limit).
- **`between`** is a **VIP List Checker** (True if you're on the list, False if not).

## 📊 Visual Diagram
```text
Data:      [10,   50,   100,   500]
clip(20, 80) -> [20,   50,   80,    80]   <- Bouncer caps them!
between(20, 80) -> [F,    T,    F,     F]   <- VIP list check
```

## 🛠️ Syntax
```python
s.astype('float')
s.between(left, right, inclusive='both')
s.clip(lower, upper)
```

## 💻 Example & Output
```python
# Banking: String to Float
transactions['amount'] = transactions['amount'].astype(float)

# Healthcare: Clipping impossible Blood Pressure
patients['bp_clipped'] = patients['bp'].clip(lower=50, upper=250)
```
**Output:**
```text
transactions['amount']: [100.5, 200.0, NaN, 400.75, 50.0]
patients['bp_clipped']: [120, 180, 110, 130, 250, 120] <- 400 became 250!
```

## 🔍 Step-by-Step Explanation
1. `astype(float)` reads the string `'100.5'` and converts it to the numeric `100.5`.
2. `clip(50, 250)` scans the `bp` column. When it sees `400`, it says "Too high!" and replaces it with the upper limit `250`.

## 🌍 Real World Example (Healthcare)
In a hospital database, sensor glitches might record a heart rate of `0` or `1000`. We use `clip(lower=30, upper=220)` to cap the data to biologically possible human limits before feeding it to an AI diagnostic model.


# 📘 The Missing Data Trinity (`isnull()`, `notnull()`, `hasnans`, `dropna()`, `fillna()`)

## 🧠 Handling NaN (Not a Number) is the most critical part of data cleaning.
- **`isnull()` / `notnull()`**: Detects missing (`NaN`) values.
- **`Series.hasnans`**: A blazing fast attribute returning a single True/False if the Series contains any NaNs.
- **`dropna()`**: Removes rows with NaNs. The how='all' parameter drops a row ONLY if every single column is NaN. how='any' (default) drops if even one column is NaN.
- **`fillna()`**: Replaces missing data with a specific value or statistic.

## 🎯 Why Do We Need It?
Algorithms like Scikit-Learn's `LinearRegression` will throw a `ValueError` if fed `NaN`s. You must decide whether to **drop** the incomplete records or **impute** (fill) them.

## 💡 Mental Model
- **`isnull`** is a **Metal Detector** (scanning for hidden holes).
- **`dropna`** is **Throwing away contaminated food** (removing the whole row).
- **`fillna`** is **Patching a hole** (replacing the missing piece).

## 📊 Visual Diagram
```text
Data: [1, NaN, 3]
isnull()   -> [False, True, False]
dropna()   -> [1, 3]
fillna(0)  -> [1, 0, 3]
```

## 🛠️ Syntax
```python
df.isnull().sum()
df.dropna(how='any' | 'all', subset=['col'])
df.fillna(value | df.mean())
```

## 💻 Example & Output
```python
# Healthcare: Fill missing ages with median
patients['age_filled'] = patients['age'].fillna(patients['age'].median())

# E-commerce: Drop orders where rating is missing
clean_orders = orders.dropna(subset=['rating'])
```
**Output:**
```text
patients['age_filled']: [25.0, 30.0, 30.0, 45.0, 30.0, 30.0] <- NaN became 30.0!
clean_orders shape: (4, 4) <- Row 3 was dropped!
```

## 🔍 Step-by-Step Explanation
1. `patients['age'].median()` calculates 30.0 (ignoring the NaN).
2. `fillna()` replaces the `NaN` in row 3 with 30.0.
3. `dropna(subset=['rating'])` looks ONLY at the `rating` column. If it sees `NaN`, it drops that entire row.

## 🌍 Real World Example (E-Commerce)
If a user buys a Laptop but doesn't leave a `rating`, we don't want to delete their entire purchase record (`dropna`). Instead, we might `fillna(0)` or `fillna(mean)` for the rating column so we can still analyze their revenue.

## ⚠️ Common Mistakes
**Wrong:**
```python
df.dropna(how='any') # Drops almost everything!
```
❌ **Logical Error:** If a 10,000-row dataset has just one random NaN in a minor column, `how='any'` deletes all 10,000 rows!
**Correct:**
```python
df.dropna(subset=['critical_column_1', 'critical_column_2'])
```

## 🎤 Interview Questions & Output Prediction
**Q1: What is the difference between `df['col'].hasnans` and `df['col'].isnull().any()`?**
**A:** Both return `True` if there is at least one NaN. However, `hasnans` is a highly optimized C-level attribute and is **much faster** for simple existence checks on large Series.

**Q2: Predict the output:**
```python

```
**A:** `(1, 2)`. 
*Explanation:* Row 0 has a value in 'A', so it's kept. Row 1 is all NaNs, so it's dropped.


# 📘 Duplicates & Uniqueness (`duplicated()`, `drop_duplicates()`, `unique()`, `nunique()`)

## 🧠 Datasets often contain repeated entries or require categorical counting.
- **`duplicated()`**: Returns a Boolean mask marking rows that are exact copies of previous rows.
- **`drop_duplicates()`**: Actually deletes the duplicate rows.
- **`unique()`**: Returns a NumPy array of all distinct values in a Series.
- **`nunique()`**: Returns the integer *count* of unique values.

## 🎯 Why Do We Need It?
Duplicate transactions in banking can lead to double-charging customers. In ML, duplicates cause data leakage, making your model perform artificially well in testing but fail in production.

## 💡 Mental Model
- **`duplicated`** is a **Highlighter** (marks the copies).
- **`drop_duplicates`** is a **Shredder** (destroys the copies).
- **`unique`** is a **Bag of Marbles** (dumps out one of each color).
- **`nunique`** is a **Counter** (just tells you how many colors are in the bag).

## 📊 Visual Diagram
```text
Data: ['A', 'B', 'A', 'C']
duplicated()      -> [F, F, T, F]
drop_duplicates() -> ['A', 'B', 'C']
unique()          -> array(['A', 'B', 'C'])
nunique()         -> 3
```

## ⚠️ Common Mistakes & Interview Trap
**The NaN Trap:** 
`nunique()` **ignores NaNs by default!** If you want to count NaN as a unique category, you must explicitly pass `dropna=False`. `unique()` **includes them**.
```python
s = pd.Series(['A', 'A', np.nan])
s.nunique() # Returns 1
s.unique()  # Returns array(['A', nan])
```

## 🎤 Interview Questions & Output Prediction
**Q: Predict the output:**
```python
df = pd.DataFrame({'A': [1, 1, 2], 'B': [3, 3, 4]})
print(df.drop_duplicates(subset=['A'], keep='last'))
```
**A:** 
```text
   A  B
1  1  3
2  2  4
```
*Explanation:* It looks only at column 'A'. For the duplicate '1', it keeps the 'last' occurrence (index 1).

# 📘 Index & Column Manipulation
**Methods Covered:** `set_index()`, `reset_index()`, `sort_index()`, `rename()`, `insert()`, `drop()`

## 🧠 The Index is the backbone of the DataFrame.
- **`set_index()`**: Promotes a column to become the row index.
- **`reset_index()`**: Demotes the index back to a standard column.
- **`sort_index()`**: Sorts data by the index labels.
- **`rename()`**: Changes column or index names.
- **`insert(loc, column, value)`**: Adds a new column at a specific position (index location) rather than at the end of the DataFrame.
- **`drop()`**: Removes columns `(use axis=1)` or rows `(use axis=0)`.

## 💡 Mental Model
- **`set_index`** is putting a **Name Tag** on a row.
- **`reset_index`** is taking the Name Tag off and putting it in a **Pocket** (a new column).
- **`insert`** is cutting into a deck of cards and sliding a new card into the exact middle.

## ⚠️ Common Mistakes
**The Lost Index Mistake:**
```python
df.set_index('new_col') # The old index is permanently deleted!
```
**Correct:**
```python
df = df.reset_index().set_index('new_col') # Saves old index as a column first!
```

## 🎤 Interview Questions & Output Prediction
**Q: Predict the output:**
```python
s = pd.Series([10, 20], index=['a', 'b']).value_counts()
print(type(s.reset_index()))
```
**A:** `<class 'pandas.core.frame.DataFrame'>`. 
*Explanation:* `value_counts()` returns a Series. Calling `reset_index()` on a Series automatically converts it into a 2-column DataFrame!

**Q: How do you convert a Pandas Series into a DataFrame?**
**A:** Whenever you run an aggregation like `value_counts()` it returns a Series. By chaining `.reset_index()` to it, Pandas automatically pushes the Series index into a column and converts the whole object into a 2D DataFrame!


# 📘 Sorting, Ranking & Extremes
**Methods Covered:** `sort_values()`, `rank()`, `nlargest()`, `nsmallest()`, `value_counts()`

## 🧠 Sorting multi-dimensional data requires precision.
- **`sort_values()`**: Sorts data by column values (supports multiple columns & NaN handling). The `na_position='first'` argument brings missing values to the top instead of the bottom.
- **`rank()`**: Assigns numerical ranks (1st, 2nd, 3rd) to values.
- **`nlargest(n, col)` / `nsmallest(n, col)`**: The fastest, most memory-efficient way to grab the top/bottom 'n' rows without sorting the entire dataset.
- **`value_counts()`**: Returns the frequency of unique values.

## 💡 Mental Model
- **`sort_values`** is organizing a **deck of cards** by suit and rank.
- **`rank`** is handing out **Medals** (Gold, Silver, Bronze) based on performance.
- **`nlargest`** is a **VIP Fastpass** (skips the line to just grab the top 3).

## ⚠️ Common Mistakes
**The NaN Sorting Mistake:**
By default, `sort_values` puts NaNs at the **end**. If you sort descending, NaNs still go to the end (which is visually confusing).
**Correct:**
```python
df.sort_values('col', ascending=False, na_position='last')
```

## 🎤 Interview Questions & Output Prediction
**Q: Why use `nlargest(5)` instead of `sort_values().head(5)`?**
**A:** `nlargest()` uses a heap queue algorithm under the hood. It is $O(N \log K)$ whereas sorting is $O(N \log N)$. For massive datasets, `nlargest` is significantly faster!

**Q: Predict the output:**
```python
s = pd.Series([10, 20, 20, 30])
print(s.rank(method='dense'))
```
**A:** `[1.0, 2.0, 2.0, 3.0]`. 
*Explanation:* 'dense' ranking leaves no gaps between ranks. The two 20s tie for 2nd, and 30 gets 3rd.

# 📘 The Swiss Army Knife
**Methods Covered:** `apply()`, `isin()`, `corr()`, `copy()`

## 🧠 
- **`apply()`**: Runs a custom Python function or `lambda` on every row/column/element.
- **`isin()`**: Filters based on a list of values.
- **`corr()`**: Calculates Pearson correlation between numeric columns.
- **`copy()`**: Creates a deep, independent copy of the data.

## 💡 Mental Model
- **`apply`** is a **Factory Conveyor Belt** (every item gets processed by a custom tool).
- **`isin`** is a **Bouncer with a Guestlist** (checks if names are on the list).
- **`corr`** is a **Relationship Counselor** (checks if two variables move together).
- **`copy`** is a **Photocopy** (modifying the photocopy doesn't ruin the original).

## ⚠️ Common Mistakes
**The SettingWithCopyWarning:**
```python
subset = df[df['A'] > 5]
subset['B'] = 10 # Triggers a massive warning!
```
**Correct:**
```python
subset = df[df['A'] > 5].copy()
subset['B'] = 10 # Safe!
```

## 🎤 Interview Questions & Output Prediction
**Q: Why is `apply()` slower than vectorized operations?**
**A:** `apply()` forces Pandas to loop through the data in pure Python, breaking the C-level vectorization. Always use `df['A'] + df['B']` instead of `apply(lambda row: row['A'] + row['B'])` if possible!

**Requirements:**
1. **Type Cast:** Ensure `age` is float.
2. **Bounds:** Clip `bp` between 50 and 250 (remove sensor glitches).
3. **Missing Data:** Fill missing `age` with the median. Drop rows where `cholesterol` is missing.
4. **Duplicates:** Remove duplicate `patient_id` records (keep the first).
5. **Index:** Set `patient_id` as the index.
6. **Feature Engineering:** Use `apply()` to create a `health_risk` column: 'High' if bp > 140 else 'Low'.
7. **Final Output:** Sort by `bp` descending and show the top 3 highest risk patients.


````python
# ==========================================
# 🏥 MINI PROJECT SOLUTION
# ==========================================
# 1. Setup fresh data
df = patients.copy()

# 2. Type Cast & Bounds
df['age'] = df['age'].astype(float)
df['bp'] = df['bp'].clip(50, 250)

# 3. Missing Data
df['age'] = df['age'].fillna(df['age'].median())
df = df.dropna(subset=['cholesterol'])

# 4. Duplicates
df = df.drop_duplicates(subset=['patient_id'], keep='first')

# 5. Index
df = df.set_index('patient_id')

# 6. Feature Engineering
df['health_risk'] = df['bp'].apply(lambda x: 'High' if x > 140 else 'Low')

# 7. Final Output
print("--- Top 3 Highest BP Patients ---")
print(df.sort_values('bp', ascending=False).head(3))
````

### 💡 Golden Rules of Data Cleaning:
1. **Always `.copy()`** when slicing data to modify later.
2. **Check `df.info()`** before and after cleaning to track memory and types.
3. **Use `subset`** in `dropna()` and `drop_duplicates()` to avoid accidental data loss.
4. **Vectorize first**, use `apply()` only as a last resort.
