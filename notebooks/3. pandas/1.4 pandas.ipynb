# 📘 Module 1: The Split-Apply-Combine Paradigm

## 🧠  Theory
The `groupby()` operation involves three main steps:
1. **Splitting**: Dividing the data into groups based on some criteria.
2. **Applying**: Applying a function to each group independently.
3. **Combining**: Combining the results into a new data structure.

## 💡  Mental Model
Think of a **Deck of Cards**. 
- **Split**: Sort the deck into 4 piles (Hearts, Diamonds, Clubs, Spades).
- **Apply**: Count the number of face cards in *each* pile.
- **Combine**: Write down the 4 counts on a piece of paper.

## 📊  Visual Diagram
```mermaid
flowchart TD
    A[Original DataFrame] --> B[Split by 'Genre']
    B --> C[Action Group]
    B --> D[Comedy Group]
    C --> E[Apply: Mean Rating]
    D --> F[Apply: Mean Rating]
    E --> G[Combine Results]
    F --> G
```

## 🛠️ Syntax
```python
df.groupby('column_name')['target_column'].aggregation_function()
```

## 💻 Example 
```python
# Find the genre with highest avg IMDB rating
avg_ratings = movies.groupby('genre')['imdb_rating'].mean()
```
**Output:**
```text
genre
Action    8.500000
Comedy    7.650000
Sci-Fi    8.150000
Name: imdb_rating, dtype: float64
```
**Explanation:** 
1. `groupby('genre')` splits the DataFrame into 3 separate chunks.
2. `['imdb_rating']` selects only the rating column from those chunks.
3. `.mean()` calculates the average for each chunk and combines them into a Series.

## 🎤  Interview Questions & Output Prediction
**Q1: What is the difference between `size()` and `count()` in GroupBy?**
> **A:** `size()` counts the number of rows in each group (including NaNs). `count()` counts the number of *non-null* values in a specific column.

**🔮 Output Prediction:**
```python
df = pd.DataFrame({'A': ['X', 'X', 'Y'], 'B': [1, np.nan, 3]})
print(df.groupby('A')['B'].size())
print(df.groupby('A')['B'].count())
```
**Answer:** `size()` returns `X: 2, Y: 1`. `count()` returns `X: 1, Y: 1` (ignores the NaN in X).

# 📘 Module 2: The `agg()` Masterclass

## 🧠 1. Theory
`groupby` creates an invisible object. You can inspect it using attributes like `size()`, `first()`, `last()`, and `get_group()`.
More importantly, the `agg()` method allows you to pass multiple aggregation functions, or apply different functions to different columns simultaneously using a dictionary.

## 💡 2. Mental Model
`*get_group('Action')` is like asking a librarian to hand you only the stack of Action movies.

Think of `agg()` as a **Multi-Tool**. Instead of asking "What is the sum?", then "What is the max?", you hand the DataFrame a list of questions and get all the answers at once.

## 📊 3. Visual Diagram
```text
DataFrame
   genre  rating  revenue
   Action 8.5     800
   Action 8.1     1000
   ↓ groupby('genre').agg({
       'rating': ['mean', 'max'],
       'revenue': 'sum'
   })
Result
        rating      revenue
        mean  max   sum
genre
Action  8.3   8.5   1800
```

## 🛠️ 4. Syntax
```python
# Dictionary approach
df.groupby('col').agg({'col1': ['mean', 'max'], 'col2': 'sum'})

# Named Aggregation approach (Cleaner)
df.groupby('col').agg(avg_col1=('col1', 'mean'), sum_col2=('col2', 'sum'))
```

## 💻 5. Example & 6. Output & 7. Explanation
```python
# Dictionary approach
result = movies.groupby('genre').agg({
    'imdb_rating': ['mean', 'max'],
    'revenue_mil': 'sum'
})
```
**Output:** Creates a DataFrame with a **MultiIndex** for columns.
**Explanation:** The dictionary maps each column to the specific functions you want to apply to it.

## ⚠️ 8. Common Mistakes & Debugging
**Mistake:** Forgetting that dictionary aggregation creates a MultiIndex column structure, which can be annoying to access later.
**🐛 Debugging Task:** How do you flatten the MultiIndex columns created by `agg()`?
*(Answer: Chain `.columns = ['_'.join(col) for col in cols]` or use Named Aggregation instead!)*

Using `get_group` carelessly: 
`movies.groupby('genre').get_group('Horror')` will throw a `KeyError` because 'Horror' doesn't exist in our mock data. Use boolean filtering `df[df['genre'] == 'Horror']` if you aren't sure a category exists.

## 🎤 9. Interview Questions & Output Prediction
**Q1: What is the difference between `agg()` and `apply()`?**
**A:** `agg()` is optimized for reduction functions (turning a group into a scalar). It works column-by-column efficiently. `apply()` passes the entire sub-DataFrame to a custom function, allowing for row-dependent logic but suffering from a performance penalty.

**Q2: What is the difference between `size()` and `count()` in GroupBy?**
**A:** `size()` counts the raw number of rows in each group (including NaNs). `count()` counts the number of non-null values in a specific column.

# 📘 Module 3: Advanced GroupBy `transform()` vs `apply()`

## 🧠 1. Theory
- **`transform()`**: Like `apply`, but **must return a Series of the same size as the group**. Used for broadcasting (e.g., subtracting the group mean from every row).
- **`apply()`**: The **Wild Card**. It can do anything, return anything, and is used for complex custom logic that changes the shape of the data.

## 💡 2. Mental Model
- **`agg`** crushes the group into a **single number** (Scalar).
- **`transform`** keeps the group's **original shape** but fills it with a calculated value (Broadcasting).
- **`apply`** is a **Factory Conveyor Belt**. It processes each group and can output a scalar, a Series, or a DataFrame.

**🌍 Real World Example**: "Find the normalized rating for every movie based on its genre." You can't use `agg` because you want to keep every movie row intact. You must use `transform` to calculate the genre's mean and subtract it from the specific movie's rating.

## 📊 3. Visual Diagram
```text
Group: [10, 20, 30] (Mean = 20)

transform(lambda x: x - x.mean())
→ [-10, 0, 10]  <-- Same shape! (Normalized)

apply(lambda x: x.max())
→ 30            <-- Crushed to scalar!
```

## 🛠️ 4. Syntax
```python
# Broadcasting a group statistic back to original rows
df['new_col'] = df.groupby('col')['target'].transform('mean')

# Applying complex custom logic
df.groupby('col').apply(custom_function)
```

## 💻 5. Example & 6. Output & 7. Explanation
```python
# Find normalized IMDB rating group wise (Z-Score)
movies['rating_normalized'] = movies.groupby('genre')['imdb_rating'] \
    .transform(lambda x: (x - x.mean()) / x.std())
```
**Output:** The original DataFrame gets a new column where every movie has its Z-score relative to its genre.
**Explanation:** `transform` calculates the mean and std for the genre, then broadcasts the Z-score calculation back to every row in that genre.

## ⚠️ 8. Common Mistakes & Debugging
**Mistake:** Using `apply` when `transform` is needed.
```python
# ❌ ValueError: Length mismatch
movies['rating_normalized'] = movies.groupby('genre')['imdb_rating'].apply(lambda x: x - x.mean())
```
**Explanation:** `apply` returns one value per group (e.g., 3 values for 3 genres). You can't assign 3 values to a DataFrame with 7 rows!

## 🎤 9. Interview Questions & Output Prediction
**Q1: Why is `groupby().apply()` slow?**
**A:** Because it breaks Pandas' C-level vectorization. It forces Pandas to loop through the groups in pure Python. Always prefer built-in `agg()` or `transform()` methods if possible!

# 📘 Module 4: Multi-Column GroupBy & IPL Deep Dive

## 🧠 1. Theory
You can group by multiple columns by passing a **list of column names**. This creates a **MultiIndex (Hierarchical Index)**.

## 💡 2. Mental Model
Grouping by `['Actor', 'Genre']` is like organizing a closet first by **Person**, and then within each person's section, organizing by **Clothing Type**.

## 📊 3. Visual Diagram
```text
groupby(['actor', 'genre'])

actor       genre
Bale        Action     1
Cameron     Sci-Fi     1
DiCaprio    Action     2
```

## 🛠️ 4. Syntax
```python
df.groupby(['col1', 'col2']).agg(...)
```

## 💻 5. Example & 6. Output & 7. Explanation
```python
# Best actor-genre combo by avg metascore
combo = movies.groupby(['actor', 'genre'])['metascore'].mean()
```
**Output:** A Series with a MultiIndex (Actor, Genre).
**Explanation:** Pandas creates a hierarchy. To use this like a normal DataFrame, you must chain `.reset_index()`.

## ⚠️ 8. Common Mistakes & Debugging
**Mistake:** The MultiIndex Trap. After grouping by multiple columns, the result has a MultiIndex. If you try to filter it like a normal DataFrame, it will fail.
**🐛 Debugging Task:** How do you flatten a MultiIndex Series back into a standard DataFrame?
*(Answer: Chain `.reset_index()` at the end of your aggregation!)*

## 🎤 9. Interview Questions & Output Prediction
**Q1: How do you handle missing values in GroupBy?**
**A:** By default, `groupby()` drops NaN keys. To keep them, use `df.groupby('col', dropna=False)`.
